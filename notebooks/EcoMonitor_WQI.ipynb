```python
# @title
# =============================================
# Sección 2 — QA/QC de datos (diagnóstico compacto) — Vista EcoMonitor
# =============================================

import re
import numpy as np
import pandas as pd
from IPython.display import display, HTML

# -------------------------------
# 0) Traer df_datos / df_sitios
# -------------------------------
import builtins
if 'df_datos' not in globals():
    if hasattr(builtins, 'df_datos'):
        df_datos = builtins.df_datos
    else:
        raise RuntimeError("No se encontró df_datos. Corré la Sección 1.")
if 'df_sitios' not in globals():
    df_sitios = builtins.df_sitios if hasattr(builtins, 'df_sitios') else pd.DataFrame(
        columns=["id","sitio","descripcion","latitud","longitud","altura_msnm"]
    )

# -------------------------------
# 1) Configuración mínima
# -------------------------------
MIN_COLS = ["sitio","fecha","Ecoli_UFC100ml"]

# Núcleo B1 por GRUPOS (≥4 grupos con dato)
# - CE_uScm o STD_mgL (alternativas)
# - DBO5_mgL o DQO_mgL (alternativas)
# Nota: Clorofila_a_ugL, ST_mgL y STS_mgL NO son núcleo
CORE_GROUPS = [
    ["pH"],
    ["OD_mgL"],
    ["Turb_NTU"],
    ["NO3_mgL"],
    ["PO4_mgL"],
    ["CE_uScm", "STD_mgL"],   # alternativas
    ["DBO5_mgL", "DQO_mgL"],  # alternativas
]
MIN_CORE_GROUPS = 4  # mínimo requerido

# Parser de números, ND y <DL (igual al tuyo)
def parse_nd_and_commas(x):
    if pd.isna(x): return (np.nan, False, None, False)
    if isinstance(x, (int,float,np.integer,np.floating)): return (float(x), False, None, False)
    t = str(x).strip()
    # coma decimal a punto
    if re.match(r"^-?\d{1,3}(?:\.\d{3})*,\d+$", t): t = t.replace(".","").replace(",",".")
    elif re.match(r"^-?\d+,\d+$", t): t = t.replace(",",".")
    # ND / n.d.
    if re.fullmatch(r"(?i)nd|n\.d\.|na|n/a|--|—", t): return (np.nan, True, None, False)
    # <DL
    m = re.match(r"^\s*<\s*([0-9]*[.,]?[0-9]+)\s*$", t)
    if m: return (np.nan, False, float(m.group(1).replace(",",".")), False)
    # número
    try: return (float(t), False, None, False)
    except: return (np.nan, False, None, True)

# ---------------- Estilo ----------------
EM_BLUE  = globals().get("EM_BLUE",  "#1F3B64")
EM_GREEN = globals().get("EM_GREEN", "#7BC043")
EM_LIGHT = globals().get("EM_LIGHT", "#F2F7F4")
STYLE_BOX = globals().get(
    "STYLE_BOX",
    f"background:{EM_LIGHT}; border-left:4px solid {EM_GREEN}; padding:10px 12px; border-radius:10px;"
)

# ---------------- QA/QC ----------------
# Columnas mínimas
missing_min = [c for c in MIN_COLS if c not in df_datos.columns]
min_ok = (len(missing_min) == 0)

# Fechas inválidas
if "fecha" in df_datos.columns:
    nat_count = int(pd.to_datetime(df_datos["fecha"], errors="coerce").isna().sum())
    fechas_msg = f"{nat_count}"
else:
    fechas_msg = "— (sin columna 'fecha')"

# Duplicados sitio+fecha
if {"sitio","fecha"}.issubset(df_datos.columns):
    dup_count = int(df_datos.duplicated(subset=["sitio","fecha"]).sum())
    dup_msg = f"{dup_count}"
else:
    dup_msg = "— (requiere 'sitio' y 'fecha')"

# Sitios en datos no listados en 'sitios'
if "sitio" in df_datos.columns and "sitio" in df_sitios.columns and len(df_sitios)>0:
    datos_sites  = set(df_datos["sitio"].astype(str).str.strip())
    sitios_sites = set(df_sitios["sitio"].astype(str).str.strip())
    unknown_sites = sorted(list(datos_sites - sitios_sites))
    unknown_msg = f"{len(unknown_sites)}"
else:
    unknown_msg = "—"

# Coordenadas
if len(df_sitios)>0 and {"latitud","longitud"}.issubset(df_sitios.columns):
    lat_miss = int(df_sitios["latitud"].isna().sum())
    lon_miss = int(df_sitios["longitud"].isna().sum())
    if lat_miss==0 and lon_miss==0:
        coord_msg = "✅ completas"
        coord_class = "ok"
    else:
        coord_msg = f"⚠️ faltantes (lat={lat_miss}, lon={lon_miss})"
        coord_class = "warn"
else:
    coord_msg = "—"
    coord_class = "na"

# Mínimo Sanitario (E. coli)
if "Ecoli_UFC100ml" in df_datos.columns:
    ecoli_valid = pd.to_numeric(df_datos["Ecoli_UFC100ml"].apply(lambda x: parse_nd_and_commas(x)[0]), errors="coerce").notna().sum()
    sanitario_ok = (ecoli_valid > 0)
    sanitario_msg = f"{'✅' if sanitario_ok else '❌'} (válidos={ecoli_valid})"
else:
    sanitario_ok = False
    sanitario_msg = "❌ (falta columna)"

# Helper: ¿columna tiene algún dato numérico válido?
def _has_data(col):
    if col not in df_datos.columns: return False
    s = pd.to_numeric(df_datos[col].apply(lambda x: parse_nd_and_commas(x)[0]), errors="coerce")
    return bool(s.notna().any())

# Núcleo FQ por grupos (≥4 grupos con dato)
core_ok_count = sum(any(_has_data(c) for c in grp) for grp in CORE_GROUPS)
ok_fq = core_ok_count >= MIN_CORE_GROUPS
fq_msg = f"{'✅' if ok_fq else '❌'} (grupos con dato={core_ok_count}/{len(CORE_GROUPS)})"

# Avisos suaves por alternativas con dato
ce_std_both  = _has_data("CE_uScm") and _has_data("STD_mgL")
dbo_dqo_both = _has_data("DBO5_mgL") and _has_data("DQO_mgL")
ce_std_msg   = "⚠️ CE y STD tienen datos (luego elegís una)" if ce_std_both else "—"
dbo_dqo_msg  = "⚠️ DBO5 y DQO tienen datos (luego elegís una)" if dbo_dqo_both else "—"

# Variables con nulos (parcial)
nul_list = []
for c in df_datos.columns:
    s = df_datos[c]
    if s.dtype == object:
        s = s.replace(r"^\s*$", np.nan, regex=True)
    total = len(s)
    miss = int(s.isna().sum())
    if total>0:
        pnull = (miss/total)*100.0
        if 0.0 < pnull < 100.0:
            nul_list.append((c, pnull))
TOP = 8
nul_summary = ""
if nul_list:
    nul_list.sort(key=lambda t: t[1], reverse=True)
    head = [f"{name} ({p:.0f}%)" for name,p in nul_list[:TOP]]
    extra = len(nul_list) - len(head)
    tail = f" … (+{extra} más)" if extra>0 else ""
    nul_summary = "; ".join(head) + tail

# Columnas 100% nulas
zero_list = []
for c in df_datos.columns:
    s = df_datos[c]
    if s.dtype == object:
        s = s.replace(r"^\s*$", np.nan, regex=True)
    if len(s)>0 and int(s.isna().sum()) == len(s):
        zero_list.append(c)
zero_msg = f"{len(zero_list)}" if zero_list else ""

# ------------- PRESENTACIÓN (HTML con estilo EcoMonitor) -------------
display(HTML(f"""
<style>
.eco-card {{
  {STYLE_BOX}
  margin: 8px 0 20px 0;
}}
.eco-title {{
  color:{EM_BLUE};
  font-weight: 600;
  font-size: 22px;
  margin: 0 0 10px 0;
}}
.eco-divider {{
  border-top: 2px solid {EM_BLUE};
  margin: 10px 0 16px 0;
  opacity: 0.05;
}}
.eco-kv {{
  display: grid;
  grid-template-columns: 280px 1fr;
  gap: 8px 16px;
  align-items: start;
  font-size: 15px;
}}
.eco-kv .k {{color:#333}}
.eco-kv .v {{color:#111;}}
.eco-pill {{
  display:inline-block; padding:2px 8px; border-radius:999px; font-weight:700; font-size:12px;
  color:#fff; background:{EM_BLUE};
}}
.eco-pill.ok {{ background:{EM_GREEN}; }}
.eco-pill.warn {{ background:#e8a317; }}
.eco-pill.na {{ background:#888; }}
.eco-foot {{
  margin-top: 14px; font-size: 13px; color:#333;
}}
</style>

<div class="eco-card">
  <div class="eco-title">Sección 2 — QA/QC de datos (diagnóstico compacto)</div>
  <div class="eco-divider"></div>
  <div class="eco-kv">
    <div class="k">Columnas mínimas</div>
    <div class="v">{'✅ Presentes' if min_ok else f'❌ faltan: {missing_min}'}</div>

    <div class="k">Fechas inválidas</div>
    <div class="v">{fechas_msg}</div>

    <div class="k">Duplicados (sitio+fecha)</div>
    <div class="v">{dup_msg}</div>

    <div class="k">Sitios en “datos” no listados en “sitios”</div>
    <div class="v">{unknown_msg}</div>

    <div class="k">Coordenadas en “sitios”</div>
    <div class="v">{coord_msg}</div>

    <div class="k">Mínimo Sanitario (E. coli)</div>
    <div class="v">{sanitario_msg}</div>

    <div class="k">Mínimo Fisicoquímico (≥{MIN_CORE_GROUPS} grupos núcleo)</div>
    <div class="v">{fq_msg}</div>

    <div class="k">CE ↔ STD (alternativas)</div>
    <div class="v">{ce_std_msg}</div>

    <div class="k">DBO5 ↔ DQO (alternativas)</div>
    <div class="v">{dbo_dqo_msg}</div>

    <div class="k">Variables con datos nulos</div>
    <div class="v">{(nul_summary if nul_summary else '—')}</div>

    <div class="k">Columnas con 100% nulos</div>
    <div class="v">{(zero_msg if zero_msg else '—')}</div>
  </div>
</div>
"""))
```

```python
# @title
# =============================================
# Sección 3 — Selección de variables por familia (UI + reglas + prerequisitos)
# =============================================
import re, numpy as np, pandas as pd, ipywidgets as w
from IPython.display import display, HTML
import builtins

# --------- Estilo y fix de scroll en la salida ---------
EM_BLUE  = globals().get("EM_BLUE",  "#135566")
EM_GREEN = globals().get("EM_GREEN", "#7BC043")
EM_LIGHT = globals().get("EM_LIGHT", "#F2F7F4")
STYLE_BOX = globals().get("STYLE_BOX", f"background:{EM_LIGHT}; border-left:4px solid {EM_GREEN}; padding:12px 14px; border-radius:12px;")

display(HTML(f"""
<style>
/* Botones con estilo EcoMonitor (mismo que Sección 1) */
.em-btn.widget-button {{
  border-radius: 12px !important;
  font-weight: 700 !important;
  border: none !important;
  width: 300px !important;
  height: 40px !important;
}}
.em-btn-blue.widget-button {{ background: {EM_BLUE} !important; color: #ffffff !important; }}
.em-btn-green.widget-button {{ background: {EM_GREEN} !important; color: #ffffff !important; }}

/* Tarjetas y tipografías */
.sel-card   {{ {STYLE_BOX} margin:10px 0 16px 0; }}
.sel-title  {{ color:{EM_BLUE}; font-weight:800; font-size:20px; margin:0 0 8px 0; }}
.sel-subtitle {{ color:{EM_BLUE}; font-weight:700; font-size:16px; margin:2px 0 6px 0; }}
.sel-small  {{ font-size:13px; line-height:1.5; }}
.sel-note   {{ font-size:12px; opacity:0.85; }}
.sel-warn   {{ font-size:13px; color:#B00020; font-weight:700; }}

/* Layout */
.sel-divider{{ border-top:2px solid {EM_BLUE}; margin:10px 0 14px 0; opacity:0.06; }}
.sel-grid   {{ display:grid; grid-template-columns: repeat(2, minmax(260px, 1fr)); gap:8px 16px; }}
.sel-row    {{ display:flex; align-items:center; gap:12px; }}

/* Quitar scroll interno del área de salida (Colab/Jupyter) */
.output_wrapper, .output, .output_area {{
  height: auto !important;
  max-height: none !important;
}}
.output_scroll {{
  box-shadow: none !important;
  -webkit-box-shadow: none !important;
  overflow-y: visible !important;
  height: auto !important;
  max-height: none !important;
}}
</style>
"""))

# --------- Protecciones / datos ---------
if 'df_datos' not in globals():
    if hasattr(builtins, 'df_datos'):
        df_datos = builtins.df_datos
    else:
        raise RuntimeError("No se encontró df_datos. Corré la Sección 1.")

def parse_nd_and_commas(x):
    if pd.isna(x): return (np.nan, False, None, False)
    if isinstance(x, (int,float,np.integer,np.floating)): return (float(x), False, None, False)
    t = str(x).strip()
    if re.match(r"^-?\d{1,3}(?:\.\d{3})*,\d+$", t): t = t.replace(".","").replace(",",".")
    elif re.match(r"^-?\d+,\d+$", t): t = t.replace(",",".")
    if re.fullmatch(r"(?i)nd|n\.d\.|na|n/a|--|—", t): return (np.nan, True, None, False)
    m = re.match(r"^\s*<\s*([0-9]*[.,]?[0-9]+)\s*$", t)
    if m: return (np.nan, False, float(m.group(1).replace(",",".")), False)
    try: return (float(t), False, None, False)
    except: return (np.nan, False, None, True)

def _has_data(col):
    if col not in df_datos.columns: return False
    s = pd.to_numeric(df_datos[col].apply(lambda x: parse_nd_and_commas(x)[0]), errors="coerce")
    return bool(s.notna().any())

# --------- Familias / grupos ---------
# B1 núcleo (7 grupos; CE↔STD y DBO5↔DQO son exclusivos)
B1_GRUPOS = [
    (["pH"],               "pH"),
    (["OD_mgL"],           "Oxígeno disuelto (OD)"),
    (["Turb_NTU"],         "Turbidez"),
    (["NO3_mgL"],          "Nitrato (NO3)"),
    (["PO4_mgL"],          "Fosfato (PO4)"),
    (["CE_uScm","STD_mgL"],"Sales"),
    (["DBO5_mgL","DQO_mgL"],"Materia orgánica"),
]
MIN_CORE_GROUPS = 4

# B1 opcionales (no cuentan al mínimo de 4)
B1_OPCIONALES = [
    (["STS_mgL"],          "Sólidos en suspensión (STS)"),
    (["ST_mgL"],           "Sólidos totales (ST)"),
    (["Clorofila_a_ugL"],  "Clorofila-a"),
    (["NH4_mgL"],          "Amonio (NH4)"),
    (["NO2_mgL"],          "Nitrito (NO2)"),
]

# B2-S sanitario
B2S = [
    (["Ecoli_UFC100ml"],   "E. coli (sanitario)"),
]

# B2-T toxicológico (alerta/veto)
B2T = [
    (["As_ugL"], "Arsénico (As)"),
    (["Pb_ugL"], "Plomo (Pb)"),
    (["Cd_ugL"], "Cadmio (Cd)"),
    (["Hg_ugL"], "Mercurio (Hg)"),
    (["F_mgL"],  "Fluoruro (F)"),
]

# Contexto / iones
CTX = [
    (["Temp_C"],          "Temperatura"),
    (["Alcalinidad_mgL"], "Alcalinidad"),
    (["Na_mgL"],          "Sodio (Na)"),
    (["K_mgL"],           "Potasio (K)"),
    (["Ca_mgL"],          "Calcio (Ca)"),
    (["Mg_mgL"],          "Magnesio (Mg)"),
    (["Cl_mgL"],          "Cloruro (Cl)"),
    (["SO4_mgL"],         "Sulfato (SO4)"),
    (["CT_UFC100ml"],     "Coliformes totales (contexto)"),
]

# --------- Widgets por variable ---------
CBX = {}  # checkboxes por variable

def _mk_checkbox(var, label, default=False):
    present = var in df_datos.columns
    hasval  = _has_data(var) if present else False
    desc = label if hasval else f"{label}  (sin datos)"
    cb = w.Checkbox(description=desc, value=(default and hasval), disabled=(not hasval))
    return cb

def _mk_row(vars_list):
    items = []
    if len(vars_list) == 1:
        v = vars_list[0]
        cb = _mk_checkbox(v, v, default=False)
        CBX[v] = cb
        items = [cb]
    else:
        a, b = vars_list
        cb_a = _mk_checkbox(a, a, default=False)
        cb_b = _mk_checkbox(b, b, default=False)
        CBX[a], CBX[b] = cb_a, cb_b
        items = [cb_a, cb_b]
    return w.HBox(items, layout=w.Layout(align_items="center", gap="12px"))

def _mk_group_box(title, groups, subtitle=True):
    rows = []
    for vars_list, _label in groups:
        rows.append(_mk_row(vars_list))
    header_cls = "sel-subtitle" if subtitle else "sel-title"
    header = w.HTML(f"<div class='{header_cls}'>{title}</div>")
    card = w.VBox([header, w.VBox(rows, layout=w.Layout(gap='6px'))], layout=w.Layout(padding="0"))
    return card

# --------- UI principal ---------
title = w.HTML("<div class='sel-title'>Sección 3 — Seleccionar variables</div>")
intro  = w.HTML(
    f"<div class='sel-card sel-small'>"
    f"<b>Cómo usar:</b> marcá las variables que querés usar en el índice."
    f" Los pares <b>CE↔STD</b> y <b>DBO5↔DQO</b> son <b>exclusivos</b>."
    f" Requisito para continuar: <b>E. coli seleccionado</b> y <b>≥{MIN_CORE_GROUPS}</b> grupos núcleo activos."
    f"</div>"
)

box_b1_core   = _mk_group_box("B1 — Núcleo (7 grupos, 2 exclusivos)", B1_GRUPOS, subtitle=True)
box_b1_opt    = _mk_group_box("B1 — Opcionales", B1_OPCIONALES, subtitle=True)
box_b2s       = _mk_group_box("B2-S — Sanitario", B2S, subtitle=True)
box_b2t       = _mk_group_box("B2-T — Toxicológico", B2T, subtitle=True)
box_ctx       = _mk_group_box("Contexto / iones", CTX, subtitle=True)

msg_core    = w.HTML("")
msg_require = w.HTML("")  # acá muestro errores de prerequisitos

# Botones (mismo estilo que Sección 1)
btn_confirm     = w.Button(description="Confirmar selección",
                           layout=w.Layout(width="300px", height="40px"))
btn_confirm.add_class("em-btn"); btn_confirm.add_class("em-btn-green")

btn_preset_core = w.Button(description="Preset: Núcleo recomendado",
                           layout=w.Layout(width="300px", height="40px"))
btn_preset_core.add_class("em-btn"); btn_preset_core.add_class("em-btn-blue")

btn_clear_all   = w.Button(description="Limpiar B1 opcionales y contexto",
                           layout=w.Layout(width="300px", height="40px"))
btn_clear_all.add_class("em-btn"); btn_clear_all.add_class("em-btn-blue")

out = w.Output()

# --------- Reglas de exclusión ---------
def _enforce_pair(a, b, changed):
    if changed is CBX[a] and CBX[a].value and CBX[b].value:
        CBX[b].value = False
    elif changed is CBX[b] and CBX[b].value and CBX[a].value:
        CBX[a].value = False

def _core_group_active(vars_list):
    return any((v in CBX and CBX[v].value) for v in vars_list)

def _core_group_available(vars_list):
    return any((v in CBX and not CBX[v].disabled) for v in vars_list)

def _count_core_active():
    return sum(1 for vars_list, _ in B1_GRUPOS if _core_group_active(vars_list))

def _count_core_available():
    return sum(1 for vars_list, _ in B1_GRUPOS if _core_group_available(vars_list))

def _ecoli_selected():
    return ("Ecoli_UFC100ml" in CBX) and CBX["Ecoli_UFC100ml"].value

def _update_summary_and_prereq():
    core_cnt = _count_core_active()
    core_avl = _count_core_available()
    msg_core.value = (
        f"<div class='sel-card sel-small'>"
        f"<b>Estado B1:</b> grupos núcleo activos = <b>{core_cnt}</b> / {len(B1_GRUPOS)} "
        f"(meta ≥ {MIN_CORE_GROUPS}; disponibles con datos = {core_avl})."
        f"</div>"
    )
    ready = _ecoli_selected() and (core_cnt >= MIN_CORE_GROUPS)
    if not _ecoli_selected():
        msg_require.value = "<div class='sel-warn'>⚠️ Debés seleccionar E. coli (B2-S).</div>"
    elif core_cnt < MIN_CORE_GROUPS:
        if core_avl >= MIN_CORE_GROUPS:
            msg_require.value = f"<div class='sel-warn'>⚠️ Activá al menos {MIN_CORE_GROUPS} grupos núcleo de B1.</div>"
        else:
            msg_require.value = (
                f"<div class='sel-warn'>⚠️ Tu dataset no tiene suficientes variables núcleo con datos "
                f"(disponibles {core_avl}). Agregá más datos/columnas en la Sección 1.</div>"
            )
    else:
        msg_require.value = ""
    btn_confirm.disabled = not ready

def on_toggle(change):
    if change.get("name") != "value": return
    if "CE_uScm" in CBX and "STD_mgL" in CBX:
        _enforce_pair("CE_uScm","STD_mgL", change["owner"])
    if "DBO5_mgL" in CBX and "DQO_mgL" in CBX:
        _enforce_pair("DBO5_mgL","DQO_mgL", change["owner"])
    _update_summary_and_prereq()

# Suscribir a todos los checkboxes
def _subscribe_all():
    for k, cb in CBX.items():
        cb.observe(on_toggle, names="value")

# --------- Defaults: preseleccionar núcleo + E. coli ---------
def _apply_defaults():
    # B1 núcleo: ON si hay datos; en pares preferimos primera (CE, DBO5)
    for vars_list, _ in B1_GRUPOS:
        if len(vars_list) == 1:
            v = vars_list[0]
            if v in CBX and not CBX[v].disabled:
                CBX[v].value = True
        else:
            a,b = vars_list
            if a in CBX and not CBX[a].disabled:
                CBX[a].value = True
                if b in CBX: CBX[b].value = False
            elif b in CBX and not CBX[b].disabled:
                CBX[b].value = True
    # B1 opcionales y contexto: OFF
    for coll in (B1_OPCIONALES, CTX):
        for vars_list, _ in coll:
            for v in vars_list:
                if v in CBX and not CBX[v].disabled:
                    CBX[v].value = False
    # B2-S: E. coli ON si hay datos
    if "Ecoli_UFC100ml" in CBX and not CBX["Ecoli_UFC100ml"].disabled:
        CBX["Ecoli_UFC100ml"].value = True

# --------- Presets / limpiar ---------
def _preset_core(_=None):
    _apply_defaults()
    _update_summary_and_prereq()

def _clear_b1opt_ctx(_=None):
    for coll in (B1_OPCIONALES, CTX):
        for vars_list, _ in coll:
            for v in vars_list:
                if v in CBX and not CBX[v].disabled:
                    CBX[v].value = False
    _update_summary_and_prereq()

btn_preset_core.on_click(_preset_core)
btn_clear_all.on_click(_clear_b1opt_ctx)

# --------- Confirmar selección (con validación fuerte) ---------
def _collect_selected(vars_pairs):
    selected = []
    for vars_list, _label in vars_pairs:
        for v in vars_list:
            if v in CBX and CBX[v].value:
                selected.append(v)
    return selected

def on_confirm(_):
    # Validación fuerte
    core_cnt = _count_core_active()
    if (not _ecoli_selected()) or (core_cnt < MIN_CORE_GROUPS):
        _update_summary_and_prereq()
        with out:
            out.clear_output()
            display(HTML("<div class='sel-card sel-warn'>❌ No podés continuar: cumplí los requisitos indicados arriba.</div>"))
        return

    sel_b1   = _collect_selected(B1_GRUPOS) + _collect_selected(B1_OPCIONALES)
    sel_b2s  = _collect_selected(B2S)
    sel_b2t  = _collect_selected(B2T)
    sel_ctx  = _collect_selected(CTX)
    alt = {
        "CE":   "CE_uScm"  in sel_b1,
        "STD":  "STD_mgL"  in sel_b1,
        "DBO5": "DBO5_mgL" in sel_b1,
        "DQO":  "DQO_mgL"  in sel_b1,
    }

    builtins.SEL = {
        "B1": sel_b1,
        "B2S": ("Ecoli_UFC100ml" in sel_b2s),
        "B2T": sel_b2t,
        "contexto": sel_ctx,
        "alternativas": alt,
    }
    globals()["SEL"] = builtins.SEL

    with out:
        out.clear_output()
        display(HTML(
            f"<div class='sel-card'>"
            f"<br><b>Selección confirmada.</b><br><br>"
        ))

btn_confirm.on_click(on_confirm)

# --------- Render (preselección + suscripción + estado inicial) ---------
# Construcción visual
controls = w.VBox([btn_preset_core, btn_clear_all, btn_confirm],
                  layout=w.Layout(align_items="flex-start"))
controls.add_class("em-mv")

# Suscribir y aplicar defaults ANTES de mostrar
_subscribe_all()
_apply_defaults()
_update_summary_and_prereq()

display(w.VBox([
    title,
    intro,
    w.VBox([box_b1_core, box_b1_opt, box_b2s, box_b2t, box_ctx], layout=w.Layout(gap="6px")),
    msg_core,
    msg_require,
    controls,
    out
], layout=w.Layout(gap="10px")))
```

En este paso configurás, por **cada variable**:

- **Valor objetivo:** referencia “buena” (p. ej., Turbidez ≤ 25 NTU).
- **Valor crítico:** referencia “mala” o “peor aceptable”. Podés:

  - usar **Manual** y escribirlo vos.
- **Peso (B1):** importancia relativa de la variable. El sistema muestra al lado el **peso normalizado** (se reescala para que la suma de pesos activos sea 1).  
  > Bloque 2 (**E. coli** y tóxicos) **no** usa pesos.

```python
# @title
# ==========================================================
# Sección 3 — Umbrales & Pesos (B1 / B2) — UI alineada (con bloque y peso_norm)
# ==========================================================
import numpy as np, pandas as pd, ipywidgets as w
from IPython.display import display, HTML
import builtins

# ---------- Estilo EcoMonitor ----------
EM_BLUE  = globals().get("EM_BLUE",  "#135566")
EM_GREEN = globals().get("EM_GREEN", "#7BC043")
EM_LIGHT = globals().get("EM_LIGHT", "#F2F7F4")
STYLE_BOX = globals().get("STYLE_BOX",
    f"background:{EM_LIGHT}; border-left:4px solid {EM_GREEN}; padding:12px 14px; border-radius:12px;"
)

display(HTML(f"""
<style>
.em-btn.widget-button {{
  border-radius: 12px !important;
  font-weight: 700 !important;
  border: none !important;
  width: 300px !important; height: 40px !important;
}}
.em-btn-blue.widget-button  {{ background: {EM_BLUE} !important; color:#fff !important; }}
.em-btn-green.widget-button {{ background: {EM_GREEN} !important; color:#fff !important; }}

.block-card {{ {STYLE_BOX} margin: 8px 0 16px 0; }}
.block-title {{ color:{EM_BLUE}; font-weight:800; font-size:18px; margin:0 0 6px 0; }}
.block-help {{ font-size:13px; line-height:1.5; opacity:.95; margin: 4px 0 10px 0; }}

.grid-head-b1, .grid-row-b1 {{
  display:grid;
  grid-template-columns: 260px 140px 140px 180px 110px 140px; /* Var | Obj | Crit | Rigor | Peso | Peso norm */
  gap: 10px; align-items:center;
}}
.grid-head-b1 .h {{ font-weight:700; color:#333; }}

.grid-head-b2, .grid-row-b2 {{
  display:grid;
  grid-template-columns: 260px 160px 160px; /* Var | Obj | Crit */
  gap: 10px; align-items:center;
}}
.grid-head-b2 .h {{ font-weight:700; color:#333; }}

/* Sin scroll interno en la salida */
.output_wrapper, .output, .output_area {{ height:auto !important; max-height:none !important; }}
.output_scroll {{ overflow-y: visible !important; height:auto !important; max-height:none !important; box-shadow:none !important; }}
</style>
"""))

# ---------- Utilidades ----------
def safe_num(x, fallback):
    try:
        xx = float(x)
        if np.isnan(xx) or np.isinf(xx): return float(fallback)
        return xx
    except Exception:
        return float(fallback)

def _norm_sentido(s):
    s = (s or "").strip().lower()
    mp = {"≤":"<=","=<":"<=", "≥":">=", "=>":">=", "range":"rango", "between":"rango"}
    return mp.get(s, s)

SPEC_DECIMALS = {"pH":1, "OD_mgL":2, "Turb_NTU":0, "PO4_mgL":3}
def infer_decimals_from_objective(var, o_val):
    if var in SPEC_DECIMALS: return SPEC_DECIMALS[var]
    O = abs(safe_num(o_val, 0.0))
    if O >= 100: return 0
    if O >= 10:  return 1
    if O >= 1:   return 2
    return 3

def round_value(var, val, o_val):
    if val is None: return None
    d = infer_decimals_from_objective(var, o_val)
    try:
        return int(round(float(val))) if d==0 else float(np.round(float(val), d))
    except Exception:
        return val

def format_value_as_str(var, val, o_val):
    if val is None or (isinstance(val, float) and np.isnan(val)): return ""
    d = infer_decimals_from_objective(var, o_val)
    try:
        fv = float(val)
        return f"{int(round(fv))}" if d==0 else f"{fv:.{d}f}"
    except Exception:
        return str(val)

# ---------- Fallbacks ----------
if 'ALL_PARAMS' not in globals():
    ALL_PARAMS = [
        ("Ecoli_UFC100ml","UFC/100mL"), ("CT_UFC100ml","UFC/100mL"),
        ("pH","adim"), ("Temp_C","°C"), ("CE_uScm","µS/cm"), ("OD_mgL","mg/L"),
        ("Turb_NTU","NTU"), ("Alcalinidad_mgL","mg/L"),
        ("Cl_mgL","mg/L"), ("SO4_mgL","mg/L"), ("Na_mgL","mg/L"), ("K_mgL","mg/L"),
        ("Ca_mgL","mg/L"), ("Mg_mgL","mg/L"), ("F_mgL","mg/L"),
        ("NO3_mgL","mg/L"), ("NO2_mgL","mg/L"), ("PO4_mgL","mg/L"), ("NH4_mgL","mg/L"),
        ("STD_mgL","mg/L"), ("STS_mgL","mg/L"), ("ST_mgL","mg/L"),
        ("DQO_mgL","mg/L"), ("DBO5_mgL","mg/L"),
        ("Clorofila_a_ugL","µg/L"),
        ("As_ugL","µg/L"), ("Pb_ugL","µg/L"), ("Cd_ugL","µg/L"), ("Hg_ugL","µg/L"),
    ]
if 'DEFAULTS' not in globals():
    DEFAULTS = {
        "Ecoli_UFC100ml": dict(sentido="<=", objetivo=126.0,  objetivo_max=410.0, peso=0.0),
        "OD_mgL":   dict(sentido=">=", objetivo=6.0,    objetivo_max=np.nan, peso=0.25),
        "Turb_NTU": dict(sentido="<=", objetivo=50.0,   objetivo_max=np.nan, peso=0.20),
        "pH":       dict(sentido="rango", objetivo=6.5, objetivo_max=8.5,    peso=0.10),
        "NO3_mgL":  dict(sentido="<=", objetivo=45.0,   objetivo_max=np.nan, peso=0.15),
        "PO4_mgL":  dict(sentido="<=", objetivo=0.1,  objetivo_max=np.nan, peso=0.10),
        "CE_uScm":  dict(sentido="<=", objetivo=1500.0, objetivo_max=np.nan, peso=0.10),
        "DBO5_mgL": dict(sentido="<=", objetivo=3.0,    objetivo_max=np.nan, peso=0.10),
        "DQO_mgL":  dict(sentido="<=", objetivo=40.0,   objetivo_max=np.nan, peso=0.00),
        "STD_mgL":  dict(sentido="<=", objetivo=500.0, objetivo_max=np.nan, peso=0.00),
        "STS_mgL":  dict(sentido="<=", objetivo=50.0,   objetivo_max=np.nan, peso=0.00),
        "ST_mgL":   dict(sentido="<=", objetivo=1500.0, objetivo_max=np.nan, peso=0.00),
        "NH4_mgL":  dict(sentido="<=", objetivo=0.5,    objetivo_max=np.nan, peso=0.00),
        "NO2_mgL":  dict(sentido="<=", objetivo=3,   objetivo_max=np.nan, peso=0.00),
        "Clorofila_a_ugL": dict(sentido="<=", objetivo=10.0, objetivo_max=np.nan, peso=0.00),
        "As_ugL":   dict(sentido="<=", objetivo=10.0,   objetivo_max=50.0,   peso=0.00),
        "Pb_ugL":   dict(sentido="<=", objetivo=10.0,   objetivo_max=25.0,   peso=0.00),
        "Cd_ugL":   dict(sentido="<=", objetivo=3.0,    objetivo_max=10.0,   peso=0.00),
        "Hg_ugL":   dict(sentido="<=", objetivo=6.0,    objetivo_max=10.0,   peso=0.00),
        "F_mgL":    dict(sentido="<=", objetivo=1.5,    objetivo_max=2.5,    peso=0.00),
    }
SEL = globals().get("SEL", getattr(builtins, "SEL", {
    "B1": [],
    "B2S": False,
    "B2T": []     # <- importante: lista vacía por defecto (no forzar tóxicos)
}))

# ---------- Helpers de metadatos ----------
UNITS = {p:u for p,u in ALL_PARAMS}
def unit_of(v): return UNITS.get(v, "")
def sense_of(v):
    s = DEFAULTS.get(v, {}).get("sentido", "")
    if s: return _norm_sentido(s)
    if v == "pH": return "rango"
    return "<="

# ---------- Selección ----------
SEL_B1  = SEL.get("B1", [])
SEL_B2S = bool(SEL.get("B2S", True))
SEL_B2T = SEL.get("B2T", [])
# No forzar tóxicos por defecto: solo lo seleccionado en la Sección de Selección
B2_LIST = list(dict.fromkeys(SEL_B2T))  # si el usuario no seleccionó nada, queda []

CORE_ORDER = ["pH","OD_mgL","Turb_NTU","NO3_mgL","PO4_mgL","CE_uScm","STD_mgL","DBO5_mgL","DQO_mgL"]
order_index = {p:i for i,p in enumerate(CORE_ORDER)}
B1_ORDERED = sorted(SEL_B1, key=lambda x: order_index.get(x, 999))

# ---------- Contenedores ----------
ROWS_B1, ROWS_B2 = {}, {}
out_save = w.Output()

RIGOR_OPTIONS = [("Estricto", 3), ("Medio", 4), ("Tolerante", 5), ("Manual", None)]

def calc_w(sense, o_val, rigor_code):
    if rigor_code is None or o_val is None: return None
    try:
        O = float(o_val); k = float(rigor_code)
        if sense == "<=":   return O * k
        elif sense == ">=": return O / k if k != 0 else None
    except Exception:
        return None

def recalc_norm_weights():
    ps = []
    for v, m in ROWS_B1.items():
        if m.get("peso") is not None:
            ps.append(safe_num(m["peso"].value, 0.0))
    s = sum(ps) if ps else 0.0
    for v, m in ROWS_B1.items():
        lbl = m.get("peso_norm_lbl")
        if lbl is None: continue
        p = safe_num(m["peso"].value, 0.0)
        lbl.value = f"{p/s:.2f}" if s>0 else "—"

def make_row_b1(v):
    sentido = sense_of(v)
    unidad  = unit_of(v)
    d = DEFAULTS.get(v, {})
    peso_def = d.get("peso", 0.10)

    if sentido == "rango":
        omin = safe_num(d.get("objetivo", 6.5), 6.5)
        omax = safe_num(d.get("objetivo_max", 8.5), 8.5)
        lbl   = w.Label(f"{v} ({unidad})", layout=w.Layout(width="260px"))
        ominw = w.FloatText(value=round_value(v, omin, omin), layout=w.Layout(width="140px"))
        omaxw = w.FloatText(value=round_value(v, omax, omax), layout=w.Layout(width="140px"))
        rigor = w.Dropdown(options=[("—", None)], value=None, layout=w.Layout(width="180px"))
        peso  = w.BoundedFloatText(value=safe_num(peso_def, 0.10), min=0, max=1, step=0.05, layout=w.Layout(width="110px"))
        pnlab = w.Label("—", layout=w.Layout(width="140px"))
        ominw.observe(lambda ch: ch["name"]=="value" and setattr(ominw,"value",round_value(v, ominw.value, ominw.value)), names="value")
        omaxw.observe(lambda ch: ch["name"]=="value" and setattr(omaxw,"value",round_value(v, omaxw.value, omaxw.value)), names="value")
        row   = w.HBox([lbl, ominw, omaxw, rigor, peso, pnlab])
        ROWS_B1[v] = dict(tipo="rango", omin=ominw, omax=omaxw, rigor=rigor, peso=peso, peso_norm_lbl=pnlab)
        peso.observe(lambda ch: ch["name"]=="value" and recalc_norm_weights(), names="value")
        return row

    O_def = safe_num(d.get("objetivo", 0.0), 0.0)
    W_def = d.get("objetivo_max", np.nan)
    init_k = 4
    W_init = safe_num(W_def if np.isfinite(W_def) else calc_w(sentido, O_def, init_k), O_def)
    W_init = round_value(v, W_init, O_def)

    lbl   = w.Label(f"{v} ({unidad})", layout=w.Layout(width="260px"))
    O     = w.FloatText(value=O_def, layout=w.Layout(width="140px"))
    W     = w.FloatText(value=W_init, disabled=True, layout=w.Layout(width="140px"))
    rigor = w.Dropdown(options=RIGOR_OPTIONS, value=init_k, layout=w.Layout(width="180px"))
    peso  = w.BoundedFloatText(value=safe_num(peso_def, 0.10), min=0, max=1, step=0.05, layout=w.Layout(width="110px"))
    pnlab = w.Label("—", layout=w.Layout(width="140px"))

    def refresh_W(*_):
        mode = rigor.value
        if mode is None:
            W.disabled = False
            W.value = round_value(v, W.value, O.value)
        else:
            W.disabled = True
            wcalc = calc_w(sentido, O.value, mode)
            if wcalc is not None:
                W.value = round_value(v, wcalc, O.value)

    rigor.observe(lambda ch: ch["name"]=="value" and refresh_W(), names="value")
    O.observe(lambda ch: ch["name"]=="value" and refresh_W(), names="value")
    W.observe(lambda ch: ch["name"]=="value" and (rigor.value is None) and setattr(W,"value",round_value(v, W.value, O.value)), names="value")
    peso.observe(lambda ch: ch["name"]=="value" and recalc_norm_weights(), names="value")
    refresh_W()

    row = w.HBox([lbl, O, W, rigor, peso, pnlab])
    ROWS_B1[v] = dict(tipo=sentido, O=O, W=W, rigor=rigor, peso=peso, peso_norm_lbl=pnlab)
    return row

def make_row_b2(v):
    sentido = sense_of(v) if v!="Ecoli_UFC100ml" else "<="
    unidad  = unit_of(v)
    d = DEFAULTS.get(v, {})
    lbl   = w.Label(f"{v} ({unidad})", layout=w.Layout(width="260px"))
    if v == "Ecoli_UFC100ml":
        O = w.FloatText(value=safe_num(d.get("objetivo",126.0),126.0), layout=w.Layout(width="160px"))
        W = w.FloatText(value=round_value(v, safe_num(d.get("objetivo_max",410.0),410.0), O.value),
                        layout=w.Layout(width="160px"))
        W.observe(lambda ch: ch["name"]=="value" and setattr(W,"value",round_value(v, W.value, O.value)), names="value")
        row = w.HBox([lbl, O, W]); ROWS_B2[v] = dict(tipo="<=", O=O, W=W); return row
    O_def = safe_num(d.get("objetivo", 0.0), 0.0)
    W_def = safe_num(d.get("objetivo_max", 0.0), 0.0)
    O = w.FloatText(value=O_def, layout=w.Layout(width="160px"))
    W = w.FloatText(value=round_value(v, W_def, O_def), layout=w.Layout(width="160px"))
    W.observe(lambda ch: ch["name"]=="value" and setattr(W,"value",round_value(v, W.value, O.value)), names="value")
    row = w.HBox([lbl, O, W]); ROWS_B2[v] = dict(tipo=sentido, O=O, W=W); return row

# ---------- BLOQUES (UI) ----------
b1_title = w.HTML(f"<div class='block-title'>Bloque 1 — Físico-químico (promedia)</div>")
b1_help  = w.HTML("<div class='block-help'>Objetivo y crítico (ajustable con <b>Rigor</b>). "
                  "El <b>peso</b> define la importancia; a la derecha ves el <b>peso normalizado</b>.</div>")
b1_head  = w.HTML(
    "<div class='grid-head-b1'>"
    "<div class='h'>Variable (unidad)</div>"
    "<div class='h'>Valor objetivo</div>"
    "<div class='h'>Valor crítico</div>"
    "<div class='h'>Rigor</div>"
    "<div class='h'>Peso</div>"
    "<div class='h'>Peso normalizado</div>"
    "</div>"
)
b1_rows = [make_row_b1(v) for v in B1_ORDERED]
b1_box  = w.VBox([b1_title, b1_help, b1_head] + b1_rows, layout=w.Layout(gap="6px")); b1_box.add_class("block-card")

b2_title = w.HTML(f"<div class='block-title'>Bloque 2 — Sanitario / Toxicológico (sin ponderación)</div>")
b2_help  = w.HTML("<div class='block-help'><b>E. coli</b>: objetivo (seguro) y crítico (no apto). "
                  "Tóxicos: objetivo marca <i>alerta</i>, crítico sugiere <i>veto</i>.</div>")
b2_head  = w.HTML(
    "<div class='grid-head-b2'>"
    "<div class='h'>Variable (unidad)</div>"
    "<div class='h'>Valor objetivo</div>"
    "<div class='h'>Valor crítico</div>"
    "</div>"
)
b2_rows = []
if SEL_B2S: b2_rows.append(make_row_b2("Ecoli_UFC100ml"))
for v in B2_LIST: b2_rows.append(make_row_b2(v))
b2_box = w.VBox([b2_title, b2_help, b2_head] + b2_rows, layout=w.Layout(gap="6px")); b2_box.add_class("block-card")

# ---------- Guardar configuración ----------
btn_save = w.Button(description="Guardar umbrales y pesos", layout=w.Layout(width="300px", height="40px"))
btn_save.add_class("em-btn"); btn_save.add_class("em-btn-green")

def on_save(_):
    rows = []

    # B1
    for v, meta in ROWS_B1.items():
        unidad = unit_of(v)
        if meta["tipo"] == "rango":
            omin = round_value(v, meta["omin"].value, meta["omin"].value)
            omax = round_value(v, meta["omax"].value, meta["omax"].value)
            peso = safe_num(meta["peso"].value, 0.10)
            rows.append(dict(parametro=v, unidad=unidad, sentido="rango",
                             objetivo=omin, objetivo_max=omax, peso=peso, bloque="B1"))
        else:
            O     = round_value(v, safe_num(meta["O"].value, 0.0), safe_num(meta["O"].value, 0.0))
            rigor = meta["rigor"].value
            W     = round_value(v, (calc_w(meta["tipo"], O, rigor) if rigor is not None else meta["W"].value), O)
            peso  = safe_num(meta["peso"].value, 0.10)
            rows.append(dict(parametro=v, unidad=unidad, sentido=meta["tipo"],
                             objetivo=O, objetivo_max=W, peso=peso, bloque="B1"))

    # B2
    if "Ecoli_UFC100ml" in ROWS_B2:
        meta = ROWS_B2["Ecoli_UFC100ml"]
        O = round_value("Ecoli_UFC100ml", safe_num(meta["O"].value, 0.0), safe_num(meta["O"].value, 0.0))
        W = round_value("Ecoli_UFC100ml", meta["W"].value, O)
        rows.append(dict(parametro="Ecoli_UFC100ml", unidad=unit_of("Ecoli_UFC100ml"),
                         sentido="<=", objetivo=O, objetivo_max=W, peso=0.0, bloque="B2"))
    for v in B2_LIST:
        meta = ROWS_B2.get(v)
        if meta is None: continue
        O = round_value(v, safe_num(meta["O"].value, 0.0), safe_num(meta["O"].value, 0.0))
        W = round_value(v, meta["W"].value, O)
        rows.append(dict(parametro=v, unidad=unit_of(v), sentido=meta["tipo"],
                         objetivo=O, objetivo_max=W, peso=0.0, bloque="B2"))

    # DataFrame final
    df = pd.DataFrame(rows, columns=["parametro","unidad","sentido","objetivo","objetivo_max","peso","bloque"])

    # peso_norm (solo B1 con peso>0)
    m_b1 = (df["bloque"].astype(str).str.upper()=="B1") & (df["peso"]>0)
    sum_pesos = float(df.loc[m_b1, "peso"].sum())
    df["peso_norm"] = np.where(m_b1 & (sum_pesos>0), df["peso"]/sum_pesos, np.nan)

    # Guardar
    builtins.ref_rec = df.copy()
    builtins.REFS    = {"recreativo": df.copy()}
    globals()["ref_rec"] = builtins.ref_rec
    globals()["REFS"]    = builtins.REFS

    # Vista previa (seleccionar -> renombrar)  <<< FIX DEL KEYERROR
    df_disp = df.copy()
    df_disp["objetivo_str"]     = [format_value_as_str(r["parametro"], r["objetivo"], r["objetivo"]) for _, r in df.iterrows()]
    df_disp["objetivo_max_str"] = [format_value_as_str(r["parametro"], r["objetivo_max"], r["objetivo"]) for _, r in df.iterrows()]
    df_disp["peso_norm"] = df_disp["peso_norm"].astype(float)

    # Selecciono con nombres originales...
    base_cols = ["parametro","bloque","sentido","unidad","objetivo_str","objetivo_max_str","peso","peso_norm"]
    df_show = df_disp[base_cols].rename(columns={
        "parametro":"Variable","bloque":"Subíndice","sentido":"Sentido","unidad":"Unidad",
        "objetivo_str":"Objetivo","objetivo_max_str":"Crítico","peso":"Peso","peso_norm":"Peso_norm"
    })

    def _rgba(hex_color, alpha=0.12):
        hex_color = hex_color.lstrip("#")
        r = int(hex_color[0:2],16); g = int(hex_color[2:4],16); b = int(hex_color[4:6],16)
        return f"rgba({r},{g},{b},{alpha})"

    def _style_rows(_df):
        styles = pd.DataFrame("", index=_df.index, columns=_df.columns)
        if "Subíndice" in _df.columns:
            j = _df.columns.get_loc("Subíndice")
            for i, v in enumerate(_df["Subíndice"].astype(str)):
                if v.upper()=="B1":
                    styles.iloc[i, j] = f"background-color:{_rgba(EM_BLUE,0.10)}; color:#0f2547; font-weight:700;"
                elif v.upper()=="B2":
                    styles.iloc[i, j] = f"background-color:{_rgba(EM_GREEN,0.12)}; color:#144d1f; font-weight:700;"
        return styles

    df_show["Peso"] = pd.to_numeric(df_show["Peso"], errors="coerce").round(2)
    df_show["Peso_norm"] = pd.to_numeric(df_show["Peso_norm"], errors="coerce").round(2)

    with out_save:
        out_save.clear_output()
        display(HTML(f"""
            <div class='block-card'>
              <div style="color:{EM_BLUE}; font-weight:800; font-size:16px; margin-bottom:6px;">Configuración guardada</div>
              <div style="font-size:13px; color:#333;">Se guardaron <b>bloque</b> (B1/B2) y <b>peso normalizado</b> para B1.</div>
            </div>
        """))
        try:
            styled = (
             df_show.style
             .hide(axis='index')
             .apply(_style_rows, axis=None)
             .background_gradient(axis=None, subset=["Peso_norm"], cmap="Greens")
             .format({"Peso": "{:.2f}", "Peso_norm": "{:.2f}"}, na_rep="—")
             .set_properties(**{'font-size':'13px','white-space':'nowrap'})
            )
            display(styled)
        except Exception:
            display(df_show)

btn_save.on_click(on_save)

# Render
box = w.VBox([b1_box, b2_box, btn_save, out_save], layout=w.Layout(gap="10px"))
display(box)
recalc_norm_weights()
```

## **SECCIÓN 4 - CALCULO DEL INDICE**


🎯 Acá se calculan los componentes del índice para cada muestra:

**Bloque 1 (físico-químico, 0–100):**  
- Se aplican **curvas lineales** por variable entre el **objetivo** (100 puntos) y el **crítico** (0 puntos).  
- El subíndice B1 es un **promedio ponderado** con **pesos normalizados por muestra** (solo entre variables activas con dato).

**Bloque 2 (sanitario / toxicológico):**  
- **E. coli**:  
  - `≤ objetivo` → **Seguro**  
  - `entre objetivo y crítico` → **Comprometido**  
  - `> crítico` → **No apto**
- **Tóxicos**: clasifican cada muestra con **ok / alerta / veto**.  
  Cualquier **veto** vuelve el estado global del Bloque 2 a **No apto**.

**Salidas principales (vistas previas + descarga en Excel):**  
- `subindices` → por muestra: **Sub_B1** (0–100) y **Estado_B2**.  
- `scores_vars` → puntajes por **variable** de B1 (0–100) con su peso.  
- `b2_detalle` → detalle sanitario/tóxico (E. coli + alertas/vetos por parámetro).

> Notas:  
> • Los puntajes se muestran redondeados en pantalla, pero se guardan con precisión completa.  
> • Si faltan datos en alguna variable, esa variable no pondera en esa muestra.

```python
# @title
# =============================================
# Sección 4 — Cálculo de subíndices por bloque (alineado con Sección 5)
# - Curvas lineales
# - B1: pesos normalizados por muestra, vista en enteros
# - B2: sanitario + tóxicos → estados: Seguro / Comprometido / No apto / Sin dato
# - Vistas previas (5 filas) + botón de descarga debajo de cada tabla (Excel)
# =============================================
import os, math
import numpy as np
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, HTML

# ---------- Chequeos de entrada ----------
if 'df_datos' not in globals():
    raise RuntimeError("No se encontró df_datos. Corré la Sección 1 y cargá el Excel.")

df = df_datos.copy()

# Preferencia: configuración guardada desde Sección 3
if 'config_b1b2' in globals():
    um = config_b1b2.copy()
else:
    # Fallback compatible con la hoja 'umbrales' clásica (sin 'bloque')
    if 'ref_rec' in globals():
        um = ref_rec.copy()
    else:
        candidates = [
            "/content/plantilla_umbrales_cargado.xlsx",
            "/content/plantilla_umbrales_defaults.xlsx",
            "/content/plantilla_umbrales.xlsx",
        ]
        um = None
        for path in candidates:
            if os.path.exists(path):
                try:
                    xls = pd.ExcelFile(path)
                    if "umbrales" in xls.sheet_names:
                        um = pd.read_excel(xls, sheet_name="umbrales")
                        break
                except Exception:
                    pass
        if um is None:
            raise RuntimeError("No se encontró configuración de umbrales/pesos. Corré la Sección 3 y guardá la configuración.")

# ---------- Estilo visual ----------
EM_BLUE  = globals().get("EM_BLUE",  "#135566")
EM_GREEN = globals().get("EM_GREEN", "#7BC043")
EM_LIGHT = globals().get("EM_LIGHT", "#F2F7F4")
STYLE_BOX = globals().get(
    "STYLE_BOX",
    f"background:{EM_LIGHT}; border-left:4px solid {EM_GREEN}; padding:10px 12px; border-radius:10px;"
)

display(HTML(f"""
<style>
.eco-card {{
  {STYLE_BOX}
  margin: 12px 0 18px 0;
}}
.eco-title {{
  color:{EM_BLUE};
  font-weight: 700;
  font-size: 22px;
  margin: 0 0 10px 0;
}}
.eco-divider {{
  border-top: 2px solid {EM_BLUE};
  margin: 10px 0 16px 0;
  opacity: 0.05;
}}
.eco-kv {{
  display: grid;
  grid-template-columns: 260px 1fr;
  gap: 8px 16px;
  align-items: start;
  font-size: 15px;
}}
.eco-kv .k {{color:#333;}}
.eco-kv .v {{color:#111;}}
.eco-table-title {{ font-weight:700; margin: 10px 0 6px 0; color:#111; }}

/* Botones estilo EcoMonitor */
.em-btn.widget-button {{
  border-radius: 12px !important;
  font-weight: 700 !important;
  border: none !important;
  width: 220px !important; height: 40px !important;
  margin-top: 16px !important;
}}
.em-btn-blue.widget-button  {{ background: {EM_BLUE} !important; color:#fff !important; }}
.em-btn-green.widget-button {{ background: {EM_GREEN} !important; color:#fff !important; }}

/* Un poco de aire a los DataFrames */
.jp-OutputArea .dataframe, .output .dataframe {{
  margin-top: 6px;
  margin-bottom: 6px;
}}

/* Evitar scroll interno en la salida de la celda */
.output_wrapper, .output, .output_area {{ height:auto !important; max-height:none !important; }}
.output_scroll {{ overflow-y: visible !important; height:auto !important; max-height:none !important; box-shadow:none !important; }}
</style>
"""))

# ---------- Normalización/validación de la configuración ----------
def _to_float(x):
    if pd.isna(x): return np.nan
    if isinstance(x, str):
        x = x.replace(",", ".").strip()
        if x == "": return np.nan
    return pd.to_numeric(x, errors="coerce")

def _norm_sentido(s):
    s = str(s).strip().lower()
    mapping = {"≤":"<=","=<":"<=", "≥":">=","=>":">=", "range":"rango","between":"rango"}
    return mapping.get(s, s)

um = um.copy()
um.columns = [c.strip().lower() for c in um.columns]
for col in ["bloque","rigurosidad","unidad","parametro","sentido","objetivo","objetivo_max","peso"]:
    if col not in um.columns:
        um[col] = np.nan

# inferir bloque si falta (E. coli y tóxicos -> B2; resto -> B1)
def _infer_bloque(row):
    p = str(row["parametro"]).strip()
    if p.lower() == "ecoli_ufc100ml":
        return "B2"
    if p in ["As_ugL","Pb_ugL","Cd_ugL","Hg_ugL","F_mgL"]:
        return "B2"
    return "B1"

um["bloque"] = um["bloque"].astype("object")
mask_blank = um["bloque"].isna() | um["bloque"].astype(str).str.strip().eq("")
um.loc[mask_blank, "bloque"] = um[mask_blank].apply(_infer_bloque, axis=1)

# tipos y limpieza
um["parametro"]    = um["parametro"].astype(str).str.strip()
um["unidad"]       = um["unidad"].astype(str).str.strip()
um["sentido"]      = um["sentido"].apply(_norm_sentido)
um["objetivo"]     = um["objetivo"].apply(_to_float)
um["objetivo_max"] = um["objetivo_max"].apply(_to_float)
um["peso"]         = um["peso"].apply(_to_float).fillna(0.0)

# ---------- Utilidades de scoring lineal ----------
def lin_score_le(x, O, W):
    """Menor es mejor: 100 en x<=O; 0 en x>=W; lineal intermedio. Si falta W, usa 2*O como softmax."""
    if pd.isna(x) or pd.isna(O): return np.nan
    if pd.isna(W):
        W = 2.0*O if O>0 else O+abs(O)+1.0
    if x <= O: return 100.0
    if x >= W: return 0.0
    return max(0.0, 100.0*(W - x) / (W - O))

def lin_score_ge(x, O, W):
    """Mayor es mejor: 100 en x>=O; 0 en x<=W; lineal intermedio. Si falta W, usa O/2 como softmin."""
    if pd.isna(x) or pd.isna(O): return np.nan
    if pd.isna(W):
        W = O/2.0
    if x >= O: return 100.0
    if x <= W: return 0.0
    return max(0.0, 100.0*(x - W) / (O - W))

def lin_score_range(x, A, B, buf_ratio=0.25):
    """100 dentro del rango [A,B]; decae linealmente a 0 fuera con un buffer proporcional."""
    if pd.isna(x) or pd.isna(A) or pd.isna(B): return np.nan
    lo, hi = float(min(A,B)), float(max(A,B))
    span = hi - lo
    if span <= 0:
        m = 0.10*(abs(lo) if abs(lo)>0 else 1.0)
        L, U = lo-m, lo+m
        if x<=L or x>=U: return 0.0
        return 100.0*(1.0 - abs(x-lo)/m)
    buf = buf_ratio * span
    L, U = lo - buf, hi + buf
    if x < L or x > U: return 0.0
    if lo <= x <= hi: return 100.0
    if x < lo: return max(0.0, 100.0*(x - L)/(lo - L))
    return max(0.0, 100.0*(U - x)/(U - hi))

# ---------- Preparar B1 ----------
um_b1 = um[(um["bloque"].astype(str).str.upper()=="B1") & (um["sentido"]!="")].copy()

# Asegurar numéricos en df para variables B1 activas
for p in um_b1["parametro"].unique():
    if p in df.columns:
        df[p] = pd.to_numeric(df[p], errors="coerce")

# Construir tabla de scores por variable (B1)
b1_rows = []
for _, r in um_b1.iterrows():
    p, s, O, Wcrit, weight = r["parametro"], r["sentido"], r["objetivo"], r["objetivo_max"], (r["peso"] or 0.0)
    if p not in df.columns:
        continue
    x = df[p]
    if s == "<=":         sc = x.apply(lambda v: lin_score_le(v, O, Wcrit))
    elif s == ">=":       sc = x.apply(lambda v: lin_score_ge(v, O, Wcrit))
    elif s == "rango":    sc = x.apply(lambda v: lin_score_range(v, O, Wcrit))
    else:                 sc = pd.Series(np.nan, index=df.index)
    b1_rows.append(pd.DataFrame({
        "id": df.get("id", pd.Series(index=df.index, dtype="object")),
        "sitio": df.get("sitio", pd.Series(index=df.index, dtype="object")),
        "fecha": df.get("fecha", pd.Series(index=df.index)),
        "parametro": p,
        "score": sc.values,
        "peso": float(weight),
    }))

scores_vars = (pd.concat(b1_rows, ignore_index=True)
               if b1_rows else
               pd.DataFrame(columns=["id","sitio","fecha","parametro","score","peso"]))

# Subíndice B1 = promedio ponderado con pesos normalizados por muestra
if scores_vars.empty:
    b1 = df[["id","sitio","fecha"]].copy()
    b1["Sub_B1"] = np.nan
else:
    rows_b1 = []
    for keys, g in scores_vars.groupby(["id","sitio","fecha"], dropna=False):
        m = (~g["score"].isna()) & (g["peso"] > 0)
        if not m.any():
            sub_b1 = np.nan
        else:
            weights = g.loc[m, "peso"].astype(float)
            scores_ = g.loc[m, "score"].astype(float)
            w_sum = weights.sum()
            w_norm = weights / w_sum if w_sum > 0 else weights
            sub_b1 = float(np.average(scores_, weights=w_norm))
        if isinstance(keys, tuple):
            idv, sit, fec = keys
        else:
            idv, sit, fec = (None, None, None)
        rows_b1.append({"id": idv, "sitio": sit, "fecha": fec, "Sub_B1": sub_b1})
    b1 = pd.DataFrame(rows_b1)

# ---------- Preparar B2 (sanitario + tóxicos) ----------
def _thr_for(param_name):
    sub = um.loc[um["parametro"].str.strip().str.lower() == param_name.lower()]
    if sub.empty:
        return ("", np.nan, np.nan)
    row = sub.iloc[0]
    return (row.get("sentido",""), _to_float(row.get("objetivo", np.nan)), _to_float(row.get("objetivo_max", np.nan)))

sense_ec, ec_obj, ec_max = _thr_for("Ecoli_UFC100ml")

def estado_ecoli(v, obj, obj_max):
    # Estados B2 EXACTOS como en Sección 5: Seguro / Comprometido / No apto / Sin dato
    if pd.isna(v): return "Sin dato"
    try: vv = float(v)
    except: return "Sin dato"
    if pd.isna(obj_max):
        if pd.notna(obj) and vv <= obj: return "Seguro"
        if pd.notna(obj) and vv  > obj: return "No apto"
        return "Sin dato"
    if pd.notna(obj) and vv <= obj: return "Seguro"
    if pd.notna(obj) and pd.notna(obj_max) and (obj < vv <= obj_max): return "Comprometido"
    if pd.notna(obj_max) and vv > obj_max: return "No apto"
    return "Sin dato"

um_b2 = um[(um["bloque"].astype(str).str.upper()=="B2") & (um["sentido"]!="")].copy()
tox_list = [p for p in um_b2["parametro"].unique() if p.lower()!="ecoli_ufc100ml"]

for p in tox_list:
    if p in df.columns:
        df[p] = pd.to_numeric(df[p], errors="coerce")

def _eval_tox(v, sentido, obj, wcrit):
    # Devuelve ok / alerta / veto / sin_dato
    if pd.isna(v): return "sin_dato"
    try: val = float(v)
    except: return "sin_dato"
    s = _norm_sentido(sentido)
    if s == "<=":
        if pd.notna(wcrit) and val > wcrit: return "veto"
        if pd.notna(obj)   and val > obj:   return "alerta"
        return "ok"
    elif s == ">=":
        if pd.notna(wcrit) and val < wcrit: return "veto"
        if pd.notna(obj)   and val < obj:   return "alerta"
        return "ok"
    elif s == "rango":
        lo, hi = (obj, wcrit)
        if pd.isna(lo) or pd.isna(hi):
            return "sin_dato"
        if lo <= val <= hi: return "ok"
        return "alerta"
    return "sin_dato"

b2_detalle = df[["id","sitio","fecha"]].copy()
if "Ecoli_UFC100ml" in df.columns:
    b2_detalle["Ecoli_UFC100ml"] = pd.to_numeric(df["Ecoli_UFC100ml"], errors="coerce")
    b2_detalle["Estado_Ecoli"] = b2_detalle["Ecoli_UFC100ml"].apply(lambda v: estado_ecoli(v, ec_obj, ec_max))
else:
    b2_detalle["Ecoli_UFC100ml"] = np.nan
    b2_detalle["Estado_Ecoli"] = "Sin dato"

alert_cols, veto_cols = [], []
for p in tox_list:
    row = um_b2.loc[um_b2["parametro"]==p].iloc[0]
    snt, obj, wcrit = row["sentido"], row["objetivo"], row["objetivo_max"]
    col_eval = f"eval_{p}"
    if p in df.columns:
        b2_detalle[p] = pd.to_numeric(df[p], errors="coerce")
        b2_detalle[col_eval] = b2_detalle[p].apply(lambda v: _eval_tox(v, snt, obj, wcrit))
    else:
        b2_detalle[p] = np.nan
        b2_detalle[col_eval] = "sin_dato"
    b2_detalle[f"alerta_{p}"] = b2_detalle[col_eval].eq("alerta")
    b2_detalle[f"veto_{p}"]   = b2_detalle[col_eval].eq("veto")
    alert_cols.append(f"alerta_{p}")
    veto_cols.append(f"veto_{p}")

def _join_params(row, flag_cols, prefix):
    return ", ".join([c.replace(prefix,"") for c in flag_cols if c in row.index and bool(row[c])])

# --- construir columnas de alertas y vetos sin multiplicar Series por 0 ---
if alert_cols:
    b2_detalle["B2_Alertas"] = b2_detalle.apply(
        lambda r: _join_params(r, alert_cols, "alerta_"), axis=1
    )
else:
    b2_detalle["B2_Alertas"] = ""

if veto_cols:
    b2_detalle["B2_Vetos"] = b2_detalle.apply(
        lambda r: _join_params(r, veto_cols, "veto_"), axis=1
    )
else:
    b2_detalle["B2_Vetos"] = ""


def _combine_b2(row):
    # Estado final de B2 EXACTO como en Sección 5
    any_veto = any(bool(row[c]) for c in veto_cols) if veto_cols else False
    if any_veto:
        return "No apto"
    ec = row.get("Estado_Ecoli", "Sin dato")
    if ec == "No apto":
        return "No apto"
    any_alert = any(bool(row[c]) for c in alert_cols) if alert_cols else False
    if ec == "Comprometido" or any_alert:
        return "Comprometido"
    if ec == "Seguro":
        return "Seguro"
    return "Sin dato"

b2_detalle["Estado_B2"] = b2_detalle.apply(_combine_b2, axis=1)

# ---------- Integrar B1 + B2 ----------
subindices = pd.merge(
    b1[["id","sitio","fecha","Sub_B1"]],
    b2_detalle[["id","sitio","fecha","Estado_B2"]],
    on=["id","sitio","fecha"], how="outer"
).sort_values(["sitio","fecha"])

# ---------- Exponer resultados a todo el cuaderno ----------
import builtins
builtins.scores_vars = scores_vars
builtins.b2_detalle  = b2_detalle
builtins.subindices  = subindices
builtins.ref_rec     = um

# ---------- Resumen / presentación ----------
n_rows    = len(df)
n_vars_b1 = um_b1["parametro"].nunique()
n_rows_b1 = subindices["Sub_B1"].notna().sum()
b2_counts = subindices["Estado_B2"].value_counts(dropna=False).to_dict()

def _card(msg_html):
    display(HTML(f"<div class='eco-card'>{msg_html}</div>"))

def _fmt_counts(d):
    # Orden EXACTO: Seguro / Comprometido / No apto / Sin dato
    order = ["Seguro","Comprometido","No apto","Sin dato"]
    chips = []
    for k in order:
        v = d.get(k, 0)
        chips.append(f"<span style='white-space:nowrap'><b>{k}:</b> {v}</span>")
    return " · ".join(chips)

pct_subb1 = (100.0 * n_rows_b1 / n_rows) if n_rows > 0 else 0.0
b2_line   = _fmt_counts(b2_counts)

_card(f"""
  <div class="eco-title">Sección 4 — Cálculo de subíndices</div>
  <div class="eco-divider"></div>
  <div style="font-size:15px; line-height:1.55">
    <ul>
      <li><b>Bloque 1 (físico-químico):</b> puntaje 0–100 por promedio ponderado (pesos normalizados por muestra).</li>
      <li><b>Bloque 2 (sanitario + tóxicos):</b> estados = <i>Seguro</i>, <i>Comprometido</i>, <i>No apto</i> o <i>Sin dato</i> (con <b>alertas</b> y <b>vetos</b> por sustancias).</li>
    </ul>
    <div class="eco-kv" style="margin-top:8px">
      <div class="k">Unidades muestrales procesadas</div>  <div class="v">{n_rows}</div>
      <div class="k">Variables usadas en Bloque 1</div>    <div class="v">{n_vars_b1}</div>
      <div class="k">Muestras con puntaje B1</div>         <div class="v">{n_rows_b1} ({pct_subb1:.1f}%)</div>
      <div class="k">Sanitario (B2): estados</div>         <div class="v">{b2_line}</div>
    </div>
  </div>
""")

# ---------- Vistas previas y botones de descarga (Excel) ----------
try:
    from google.colab import files
    _COLAB = True
except Exception:
    _COLAB = False

def _preview(df_, n=5, int_cols=None):
    d = df_.head(n).copy()
    if int_cols:
        for c in int_cols:
            if c in d.columns:
                d[c] = pd.to_numeric(d[c], errors="coerce").round(0).astype("Int64")
    try:
        display(d.style.hide(axis='index'))
    except Exception:
        display(d)

def _save_excel(path, df_):
    try:
        import openpyxl  # ensure engine is present
        with pd.ExcelWriter(path, engine="openpyxl", mode="w") as writer:
            df_.to_excel(writer, index=False)
        if _COLAB:
            files.download(path)
        else:
            _card(f"Archivo exportado en <code>{path}</code>")
    except Exception as e:
        _card(f"❌ No se pudo exportar Excel: {e}")

# --- subindices ---
display(HTML("<div class='eco-table-title'><br><br>Tabla subindices</div>"))
_preview(subindices, n=5, int_cols=["Sub_B1"])
btn_sub = widgets.Button(description="📥 Descargar subindices.xlsx")
btn_sub.add_class("em-btn"); btn_sub.add_class("em-btn-blue")
btn_sub.on_click(lambda _ : _save_excel("/content/subindices.xlsx", subindices))
display(btn_sub)

# --- scores_vars ---
display(HTML("<div class='eco-table-title'><br>Tabla puntajes Bloque 1 (Físico-químico)</div>"))
_preview(scores_vars, n=5, int_cols=["score"])
btn_scores = widgets.Button(description="📥 Descargar scores_vars.xlsx")
btn_scores.add_class("em-btn"); btn_scores.add_class("em-btn-blue")
btn_scores.on_click(lambda _ : _save_excel("/content/scores_vars.xlsx", scores_vars))
display(btn_scores)

# --- b2_detalle ---
display(HTML("<div class='eco-table-title'><br>Tabla Bloque 2 — estados (Seguro/Comprometido/No apto/Sin dato) y detalles</div>"))
_preview(b2_detalle, n=5, int_cols=None)
btn_b2 = widgets.Button(description="📥 Descargar b2_detalle.xlsx")
btn_b2.add_class("em-btn"); btn_b2.add_class("em-btn-blue")
btn_b2.on_click(lambda _ : _save_excel("/content/b2_detalle.xlsx", b2_detalle))
display(btn_b2)
```

## **SECCIÓN 5 — INTEGRACIÓN DE BLOQUES Y MATRIZ DE DECISIÓN**


🎯 En esta parte se combinan los resultados del **Bloque 1 (físico-químico)** y del **Bloque 2 (sanitario)** para obtener una **clasificación final única** por muestra/sitio. Además, se aplica un **veto** de seguridad ante excedencias críticas.

<br>

### **Entradas**
- **Sub_B1 (0–100)**: puntaje físico-químico calculado con funciones lineales y **pesos normalizados** (Σ=1 sobre variables activas con dato).
- **Estado_B2 (E. coli)**:  
  - *Seguro* si `E. coli ≤ objetivo`.  
  - *Comprometido* si `objetivo < E. coli ≤ crítico`.  
  - *No apto* si `E. coli > crítico`.  
  - Si no hay “crítico”, se usa un esquema binario (*Seguro / No apto*).

<br>

### **Proceso**

<br>

1. **Clasificación de B1**  
   - *Buena*: Sub_B1 ≥ 80  
   - *Moderada*: 50 ≤ Sub_B1 < 80  
   - *Mala*: Sub_B1 < 50  

<br>

2. **Matriz de decisión (doble entrada)**  
   Combina la clase de B1 con el estado sanitario de B2:

   | B1 \\ B2     | Seguro       | Comprometido | No apto      | Sin dato       |
   |--------------|--------------|--------------|--------------|----------------|
   | **Buena**    | Apta         | Condicional  | No apta      | Indeterminada  |
   | **Moderada** | Aceptable    | Condicional  | No apta      | Indeterminada  |
   | **Mala**     | Mala         | No apta      | No apta      | Indeterminada  |
   | **Sin dato** | Indeterminada| Indeterminada| No apta      | Indeterminada  |

<br>

3. **Veto (seguridad)**  
   Se fuerza **No apta** si ocurre cualquiera de estas condiciones:
   - **E. coli** > **valor crítico** en la muestra.
   - Cualquier parámetro activo supera su **valor crítico**:
     - Para `≤`: valor **> crítico**.
     - Para `≥`: valor **< crítico**.
     - Para **rango**: valor **fuera** del intervalo `[objetivo, crítico]`.
   - (Opcional) Otros disparadores críticos que hayas configurado en los umbrales.

<br>

### **Salidas**
- **Clasificación final** por muestra:  
  ✅ *Apta* · ✅ *Aceptable* · ⚠️ *Condicional* · ⚠️ *Mala* · ❌ *No apta* · ❓ *Indeterminada*.  
- **Motivo de veto** (si aplica) para trazabilidad.
- Tablas de resumen y archivo descargable con la clasificación.

> **Tip:** Documentá en la Sección 3 los **objetivos** (mejor desempeño) y los **críticos** (peor aceptable) usados. Si una variable tiene comportamiento “mejor-si-menor” o “mejor-si-mayor”, eso ya queda implícito en su *sentido* (≤, ≥ o rango).

<br>

```python
# @title
# =============================================
# Sección 5 — Integración de bloques y matriz de decisión (UI EcoMonitor)
# Requiere:
#   - subindices (Sección 4): [id, sitio, fecha, Sub_B1, Estado_B2]
#   - ref_rec (Sección 3) en memoria o en disco (auto-load)
#   - df_datos (Sección 1) para evaluar veto con valores crudos
# Salida:
#   - df_clasificacion  (también exportable a Excel)
# =============================================
import os
import numpy as np
import pandas as pd
import ipywidgets as w
from IPython.display import display, HTML

# Descarga en Colab (si aplica)
try:
    from google.colab import files
    _COLAB = True
except Exception:
    _COLAB = False

# ------------------------------
# Chequeos mínimos
# ------------------------------
if 'subindices' not in globals():
    raise RuntimeError("No se encontró 'subindices'. Corré la Sección 4.")
if 'df_datos' not in globals():
    raise RuntimeError("No se encontró 'df_datos'. Corré la Sección 1.")

si  = subindices.copy()
raw = df_datos.copy()

# ------------------------------
# Paleta / estilos
# ------------------------------
EM_BLUE  = globals().get("EM_BLUE",  "#1F3B64")
EM_GREEN = globals().get("EM_GREEN", "#7BC043")
EM_LIGHT = globals().get("EM_LIGHT", "#F2F7F4")
STYLE_BOX = globals().get(
    "STYLE_BOX",
    f"background:{EM_LIGHT}; border-left:4px solid {EM_GREEN}; padding:10px 12px; border-radius:10px;"
)

display(HTML(f"""
<style>
.eco-card {{
  {STYLE_BOX}
  margin: 14px 0 18px 0;
}}
.eco-title {{
  color:{EM_BLUE};
  font-weight: 800;
  font-size: 22px;
  margin: 0 0 8px 0;
}}
.eco-subtitle {{
  font-weight:700; color:#222; margin: 4px 0 8px 0;
}}
.eco-divider {{
  border-top: 2px solid {EM_BLUE};
  margin: 8px 0 12px 0;
  opacity: 0.06;
}}
.eco-kv {{
  display: grid;
  grid-template-columns: 260px 1fr;
  gap: 8px 16px;
  align-items: start;
  font-size: 15px;
}}
.eco-kv .k {{color:#333;}}
.eco-kv .v {{color:#111;}}

.eco-badges {{
  display:flex; gap:10px; flex-wrap:wrap; margin-top:8px;
}}
.badge {{
  padding:6px 10px; border-radius:999px; font-weight:700; color:#fff; font-size:13px;
}}
.badge-apt   {{ background:#2e7d32; }}
.badge-acc   {{ background:#558b2f; }}
.badge-cond  {{ background:#f9a825; }}
.badge-mala  {{ background:#ef6c00; }}
.badge-no    {{ background:#c62828; }}
.badge-ind   {{ background:#6d6d6d; }}

.eco-table-title {{ font-weight:800; margin: 8px 0 6px 0; color:#111; }}
.eco-hint {{ font-size:13px; color:#333; }}

/* Matriz de decisión */
.eco-matrix {{
  border-collapse: collapse; margin-top:8px; font-size:14px;
}}
.eco-matrix th, .eco-matrix td {{
  border: 1px solid #ddd; padding: 6px 8px; text-align:center;
}}
.eco-matrix th {{
  background:#f7f9fb; color:#333;
}}

/* Sin scroll interno en la salida */
.output_wrapper, .output, .output_area {{ height:auto !important; max-height:none !important; }}
.output_scroll {{ overflow-y: visible !important; height:auto !important; max-height:none !important; box-shadow:none !important; }}

/* Botones */
.em-btn.widget-button {{
  border-radius: 12px !important;
  font-weight: 700 !important;
  border: none !important;
  width: 300px !important; height: 40px !important;
  margin: 16px !important;
}}
.em-btn-blue.widget-button  {{ background: {EM_BLUE} !important; color:#fff !important; }}
.em-btn-green.widget-button {{ background: {EM_GREEN} !important; color:#fff !important; }}
</style>
"""))

def _card(msg_html): display(HTML(f"<div class='eco-card'>{msg_html}</div>"))

# ------------------------------
# Utilidades / umbrales (carga/normalización)
# ------------------------------
REQUIRED_COLS_UM = ["parametro","unidad","sentido","objetivo","objetivo_max","peso"]

def _to_float(x):
    if pd.isna(x): return np.nan
    if isinstance(x, str):
        x = x.replace(",", ".").strip()
        if x == "": return np.nan
    return pd.to_numeric(x, errors="coerce")

def _norm_sentido(s):
    s = str(s or "").strip().lower()
    mapping = {"≤":"<=","=<":"<=","≥":">=","=>":">=","range":"rango","between":"rango"}
    return mapping.get(s, s)

def _infer_bloque_row(row):
    p = str(row["parametro"]).strip()
    if p.lower() == "ecoli_ufc100ml": return "B2"
    if p in ["As_ugL","Pb_ugL","Cd_ugL","Hg_ugL","F_mgL"]: return "B2"
    return "B1"

def _validate_ref_sheet(df_um):
    cols = [c.strip().lower() for c in df_um.columns]
    if not set(REQUIRED_COLS_UM).issubset(set(cols)):
        miss = [c for c in REQUIRED_COLS_UM if c not in cols]
        raise ValueError(f"Umbrales: faltan columnas {miss}")
    dd = df_um.copy()
    dd.columns = [c.strip().lower() for c in dd.columns]
    dd["parametro"]    = dd["parametro"].astype(str).str.strip()
    dd["unidad"]       = (dd["unidad"].astype(str) if "unidad" in dd.columns else "").astype(str).str.strip()
    dd["sentido"]      = dd["sentido"].apply(_norm_sentido)
    dd["objetivo"]     = dd["objetivo"].apply(_to_float)
    dd["objetivo_max"] = dd["objetivo_max"].apply(_to_float)
    dd["peso"]         = dd["peso"].apply(_to_float)
    if "bloque" not in dd.columns: dd["bloque"] = np.nan
    mask_blank = dd["bloque"].isna() | dd["bloque"].astype(str).str.strip().eq("")
    dd.loc[mask_blank, "bloque"] = dd[mask_blank].apply(_infer_bloque_row, axis=1)
    return dd

def _try_load_ref_from_disk():
    for path in ["/content/plantilla_umbrales_cargado.xlsx",
                 "/content/plantilla_umbrales.xlsx"]:
        if os.path.exists(path):
            try:
                xls = pd.ExcelFile(path)
                if "umbrales" in xls.sheet_names:
                    um_ = pd.read_excel(xls, sheet_name="umbrales")
                    return _validate_ref_sheet(um_)
            except Exception as e:
                print(f"⚠️ No se pudo leer '{path}': {e}")
    return None

# Obtener ref_rec
if 'ref_rec' in globals():
    um = _validate_ref_sheet(ref_rec.copy())
else:
    um = _try_load_ref_from_disk()
    if um is None:
        raise RuntimeError(
            "No se encontró 'ref_rec' ni archivos de umbrales en disco.\n"
            "Volvé a la Sección 3 y elegí: 'Usar archivo subido' o 'Usar valores por defecto'."
        )

# Extra: obtener umbrales de E. coli desde la configuración
ec_row = um.loc[um["parametro"].str.strip().str.lower()=="ecoli_ufc100ml"]
if not ec_row.empty:
    ec_obj  = _to_float(ec_row.iloc[0]["objetivo"])
    ec_max  = _to_float(ec_row.iloc[0]["objetivo_max"])
    # criterio crítico: si hay objetivo_max lo usamos; si no, usamos el objetivo
    ec_crit = ec_max if pd.notna(ec_max) else ec_obj
else:
    ec_obj = ec_max = ec_crit = np.nan  # sin hardcode; si falta, no aplicamos veto por E. coli aquí

# Lista de B2 (excepto E. coli)
um_b2 = um[(um["bloque"].astype(str).str.upper()=="B2") & (um["sentido"]!="")].copy()
tox_list = [p for p in um_b2["parametro"].unique() if str(p).strip().lower()!="ecoli_ufc100ml"]

# ------------------------------
# Clasificación Bloque 1 (desde Sub_B1)
# ------------------------------
def clase_b1(v):
    if pd.isna(v): return "Sin dato"
    if v >= 80:    return "Buena"
    if v >= 50:    return "Moderada"
    return "Mala"

si = si.copy()
si["Clase_B1"] = si["Sub_B1"].apply(clase_b1)

def _norm_b2(x):
    s = str(x).strip().lower()
    if "seguro" in s:        return "Seguro"
    if "compromet" in s:     return "Comprometido"
    if "no apto" in s:       return "No apto"
    if "sin dato" in s:      return "Sin dato"
    return "Sin dato"

si["Estado_B2"] = si["Estado_B2"].apply(_norm_b2)

# ------------------------------
# Matriz de decisión
# ------------------------------
MATRIZ = {
    "Buena":    {"Seguro":"Apta","Comprometido":"Condicional","No apto":"No apta","Sin dato":"Indeterminada"},
    "Moderada": {"Seguro":"Aceptable","Comprometido":"Condicional","No apto":"No apta","Sin dato":"Indeterminada"},
    "Mala":     {"Seguro":"Mala","Comprometido":"No apta","No apto":"No apta","Sin dato":"Indeterminada"},
    "Sin dato": {"Seguro":"Indeterminada","Comprometido":"Indeterminada","No apto":"No apta","Sin dato":"Indeterminada"}
}
def integrar_matriz(cl_b1, st_b2): return MATRIZ.get(cl_b1, {}).get(st_b2, "Indeterminada")

# ------------------------------
# Veto final (SOLO B2 + E. coli usando umbrales)
# ------------------------------
key_cols = [c for c in ["id","sitio","fecha"] if c in raw.columns]
if not key_cols:
    raise RuntimeError("df_datos no tiene ninguna de las columnas clave ('id','sitio','fecha').")

raw_key = raw.set_index(key_cols, drop=False)

def _get_val(row, param):
    key = tuple(row.get(k) for k in key_cols)
    try:
        r = raw_key.loc[key]
    except KeyError:
        return np.nan
    if isinstance(r, pd.DataFrame):
        r = r.iloc[0]
    return _to_float(r.get(param, np.nan))

def aplicar_veto(row):
    # E. coli: usa el crítico definido en umbrales (ec_crit). Si no existe, no vetamos por E. coli aquí.
    if pd.notna(ec_crit):
        ec = _get_val(row, "Ecoli_UFC100ml")
        if not pd.isna(ec) and ec > ec_crit:
            return True, f"E. coli > crítico ({ec} > {ec_crit})"

    # Veto SOLO por parámetros del Bloque 2 (tóxicos u otros B2), nunca por B1
    for p in tox_list:
        ur = um_b2.loc[um_b2["parametro"]==p].iloc[0]
        sentido = _norm_sentido(ur["sentido"])
        obj     = _to_float(ur["objetivo"])
        obj_max = _to_float(ur["objetivo_max"])
        val     = _get_val(row, p)
        if pd.isna(val):
            continue

        if sentido == "<=" and pd.notna(obj_max) and val > obj_max:
            return True, f"{p} > crítico ({val} > {obj_max})"
        if sentido == ">=" and pd.notna(obj_max) and val < obj_max:
            return True, f"{p} < crítico ({val} < {obj_max})"
        if sentido == "rango" and pd.notna(obj) and pd.notna(obj_max) and not (obj <= val <= obj_max):
            return True, f"{p} fuera de rango [{obj}, {obj_max}] (= {val})"
    return False, ""

# ------------------------------
# Integración final
# ------------------------------
rows = []
for _, r in si.iterrows():
    cat_b1 = r.get("Clase_B1")
    est_b2 = r.get("Estado_B2")
    base   = integrar_matriz(cat_b1, est_b2)

    veto, motivo = aplicar_veto(r)
    final = "No apta" if veto or est_b2 == "No apto" else base

    out = {
        "Sub_B1": r.get("Sub_B1"),
        "Clase_B1": cat_b1,
        "Estado_B2": est_b2,
        "Veto": bool(veto),
        "Motivo_veto": motivo,
        "Clasificacion_final": final
    }
    for k in ["id","sitio","fecha"]:
        if k in si.columns:
            out[k] = r.get(k)
    rows.append(out)

df_clasificacion = pd.DataFrame(rows)

# Traer 'serie' desde df_datos si existe
if "serie" in raw.columns:
    merge_keys = [k for k in ["id","sitio","fecha"] if k in df_clasificacion.columns and k in raw.columns]
    if merge_keys:
        df_clasificacion = df_clasificacion.merge(
            raw[merge_keys + ["serie"]].drop_duplicates(),
            on=merge_keys, how="left"
        )

# Orden/sort por sitio-fecha si están disponibles
sort_keys = [k for k in ["sitio","fecha"] if k in df_clasificacion.columns]
if sort_keys:
    df_clasificacion = df_clasificacion.sort_values(sort_keys).reset_index(drop=True)

# ------------------------------
# Presentación (tarjetas EcoMonitor)
# ------------------------------
n_reg = df_clasificacion.shape[0]
conteo_final = df_clasificacion["Clasificacion_final"].value_counts(dropna=False)

# Texto dinámico para el hint de veto por E. coli
if pd.notna(ec_crit):
    ec_hint = f"si <i>E. coli</i> &gt; {ec_crit:g} (umbral crítico definido en Sección 3)"
else:
    ec_hint = "si <i>E. coli</i> supera su umbral crítico definido en Sección 3"

# Título + resumen
_card(f"""
  <div class="eco-title">Sección 5 — Integración y clasificación final</div>
  <div class="eco-divider"></div>
  <div style="font-size:15px; line-height:1.55;">
    Combinamos el estado <b>fisicoquímico</b> (Bloque 1) con el estado <b>sanitario</b> (Bloque 2),
    aplicamos la <b>matriz de decisión</b> y, por seguridad, un <b>criterio de veto</b> exclusivo de <b>Bloque 2</b> (incluye E. coli).
  </div>
  <div class="eco-kv" style="margin-top:8px;">
    <div class="k">Registros clasificados</div> <div class="v">{n_reg}</div>
  </div>
  <div class="eco-badges">
    <span class="badge badge-apt">Apta: {int(conteo_final.get('Apta',0))}</span>
    <span class="badge badge-acc">Aceptable: {int(conteo_final.get('Aceptable',0))}</span>
    <span class="badge badge-cond">Condicional: {int(conteo_final.get('Condicional',0))}</span>
    <span class="badge badge-mala">Mala: {int(conteo_final.get('Mala',0))}</span>
    <span class="badge badge-no">No apta: {int(conteo_final.get('No apta',0))}</span>
    <span class="badge badge-ind">Indeterminada: {int(conteo_final.get('Indeterminada',0))}</span>
  </div>
""")

# Matriz de decisión (mostrada)
_card(f"""
  <div class="eco-subtitle">Cómo se decide (matriz de doble entrada)</div>
  <table class="eco-matrix">
    <tr>
      <th>Bloque 1 \\ Bloque 2</th>
      <th>Seguro</th>
      <th>Comprometido</th>
      <th>No apto</th>
      <th>Sin dato</th>
    </tr>
    <tr><th>Buena</th>     <td>Apta</td>        <td>Condicional</td> <td>No apta</td>   <td>Indeterminada</td></tr>
    <tr><th>Moderada</th>  <td>Aceptable</td>   <td>Condicional</td> <td>No apta</td>   <td>Indeterminada</td></tr>
    <tr><th>Mala</th>      <td>Mala</td>        <td>No apta</td>     <td>No apta</td>   <td>Indeterminada</td></tr>
    <tr><th>Sin dato</th>  <td>Indeterminada</td><td>Indeterminada</td><td>No apta</td><td>Indeterminada</td></tr>
  </table>
  <div class="eco-hint" style="margin-top:6px;">
    <b>Veto (solo B2):</b> {ec_hint} o algún parámetro de <b>Bloque 2</b> en condición crítica,
    la muestra pasa a <b>No apta</b> sin importar la matriz.
  </div>
""")

# ------------------------------
# Tabla: vista previa (sitio/fecha/serie primero) + descarga Excel
# ------------------------------
display(HTML("<div class='eco-table-title'><br><br>Clasificación final — vista previa (12 filas)</div>"))

# Reordenar para que id/sitio/fecha/serie vayan primero (si existen)
first_cols = [c for c in ["id","sitio","fecha","serie"] if c in df_clasificacion.columns]
rest_pref  = [c for c in ["Sub_B1","Clase_B1","Estado_B2","Veto","Clasificacion_final","Motivo_veto"] if c in df_clasificacion.columns]
others    = [c for c in df_clasificacion.columns if c not in first_cols + rest_pref]
ordered_cols = first_cols + rest_pref + others

_preview = df_clasificacion[ordered_cols].head(12).copy()
if "fecha" in _preview.columns:
    _preview["fecha"] = pd.to_datetime(_preview["fecha"], errors="coerce").dt.date

try:
    styled = (_preview
              .assign(Sub_B1=lambda d: pd.to_numeric(d.get("Sub_B1"), errors="coerce").round(0) if "Sub_B1" in d.columns else d.get("Sub_B1"))
              .style.hide(axis='index')
              .format({"Sub_B1": "{:.0f}"} if "Sub_B1" in _preview.columns else {})
             )
    display(styled)
except Exception:
    if "Sub_B1" in _preview.columns:
        _preview["Sub_B1"] = pd.to_numeric(_preview["Sub_B1"], errors="coerce").round(0)
    display(_preview)

# Botón de descarga
btn_xlsx = w.Button(description="Descargar clasificación (Excel)", layout=w.Layout(width="300px", height="40px"))
btn_xlsx.add_class("em-btn"); btn_xlsx.add_class("em-btn-green")

def _download_xlsx(_):
    path = "/content/clasificacion_wqi.xlsx"
    with pd.ExcelWriter(path, engine="openpyxl") as writer:
        df_clasificacion[ordered_cols].to_excel(writer, sheet_name="clasificacion", index=False)
    if _COLAB:
        try:
            files.download(path)
        except Exception as e:
            _card(f"⚠️ No se pudo iniciar la descarga automática. Descargá el archivo desde el panel de archivos. Detalle: {e}")
    else:
        _card(f"Archivo guardado en <code>{path}</code>")

btn_xlsx.on_click(_download_xlsx)
display(btn_xlsx)

# Exponer variable global
import builtins
builtins.df_clasificacion = df_clasificacion
```

```python
# @title
# =============================================
# 6.1 — Series temporales por sitio (UI EcoMonitor)
#   - Sub_B1 (línea 0–100)
#   - E. coli (línea; escala log; guías desde Sección 3 o fallback 126/410)
# =============================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from IPython.display import HTML, display

# ---------- Paleta / estilos (hereda si ya existen) ----------
EM_BLUE  = globals().get("EM_BLUE",  "#1F3B64")
EM_GREEN = globals().get("EM_GREEN", "#7BC043")
EM_LIGHT = globals().get("EM_LIGHT", "#F2F7F4")
STYLE_BOX = globals().get(
    "STYLE_BOX",
    f"background:{EM_LIGHT}; border-left:4px solid {EM_GREEN}; padding:10px 12px; border-radius:10px;"
)

display(HTML(f"""
<style>
.eco-card {{
  {STYLE_BOX}
  margin: 12px 0 16px 0;
}}
.eco-title {{
  color:{EM_BLUE}; font-weight:800; font-size:20px; margin:0 0 6px 0;
}}
.eco-subtitle {{
  font-weight:700; color:#222; margin:6px 0 6px 0;
}}
.eco-divider {{
  border-top: 2px solid {EM_BLUE}; margin: 6px 0 10px 0; opacity:0.06;
}}
.eco-hint {{ font-size:13px; color:#333; }}
.eco-badge {{
  display:inline-block; padding:4px 10px; border-radius:999px; background:#e8f5e9;
  color:#1b5e20; font-weight:700; font-size:12px; margin-right:6px; margin-top:4px;
}}
</style>
"""))

def _card(html): display(HTML(f"<div class='eco-card'>{html}</div>"))

# ---------- Fuentes de datos (compatibilidad) ----------
# subs ~ subindices ; datos ~ df_datos
if 'subs' not in globals():
    if 'subindices' in globals():
        subs = subindices.copy()
    else:
        raise RuntimeError("No se encontró 'subs' ni 'subindices'. Corré la Sección 4.")
if 'datos' not in globals():
    if 'df_datos' in globals():
        datos = df_datos.copy()
    else:
        raise RuntimeError("No se encontró 'datos' ni 'df_datos'. Corré la Sección 1.")

# Asegurar tipos
if 'fecha' in subs.columns:
    subs['fecha'] = pd.to_datetime(subs['fecha'], errors='coerce')
if 'fecha' in datos.columns:
    datos['fecha'] = pd.to_datetime(datos['fecha'], errors='coerce')

# ---------- Umbrales sanitarios desde Sección 3 (fallback 126/410) ----------
def _to_float(x):
    if pd.isna(x): return np.nan
    try:
        return float(str(x).replace(",", "."))
    except Exception:
        return np.nan

def _get_ecoli_thresholds():
    obj, crit = np.nan, np.nan
    um = None
    if 'config_b1b2' in globals():
        um = config_b1b2.copy()
    elif 'ref_rec' in globals():
        um = ref_rec.copy()
    if um is not None:
        try:
            df = um.copy()
            df.columns = [c.strip().lower() for c in df.columns]
            row = df.loc[df['parametro'].astype(str).str.strip().str.lower()=='ecoli_ufc100ml'].iloc[0]
            obj  = _to_float(row.get('objetivo', np.nan))
            crit = _to_float(row.get('objetivo_max', np.nan))
        except Exception:
            pass
    # Fallbacks
    if pd.isna(obj):  obj  = 126.0
    if pd.isna(crit): crit = 410.0
    return float(obj), float(crit)

EC_OBJ, EC_CRIT = _get_ecoli_thresholds()

# ---------- Sitios con ≥2 fechas ----------
series_candidates = subs.groupby("sitio")["fecha"].nunique(dropna=True)
multi_series_sites = sorted(series_candidates[series_candidates > 1].index.tolist())

# ---------- Tarjeta de resumen ----------
if len(multi_series_sites) > 0:
    _card(f"""
      <div class="eco-title">6.1 — Series temporales por sitio</div>
      <div class="eco-divider"></div>
      <div class="eco-hint">
        Se grafican los sitios con <b>al menos dos fechas</b>. Para cada sitio se muestra:
        <ul style="margin:6px 0 0 18px;">
          <li><b>Sub_B1</b> (índice fisicoquímico; escala 0–100).</li>
          <li><i>E. coli</i> (UFC/100 mL; escala log). Líneas guía: <b>{int(EC_OBJ)}</b> y <b>{int(EC_CRIT)}</b>.</li>
        </ul>
      </div>
      <div style="margin-top:8px;">{"".join([f"<span class='eco-badge'>Sitio: {s}</span>" for s in multi_series_sites])}</div>
    """)
else:
    _card(f"""
      <div class="eco-title">6.1 — Series temporales por sitio</div>
      <div class="eco-divider"></div>
      No hay sitios con dos o más fechas para trazar series temporales.
      <div class="eco-hint" style="margin-top:6px;">Cargá más campañas o periodos en la hoja <code>datos</code> para ver evolución temporal.</div>
    """)

# ---------- Función de graficado ----------
def _fmt_dates(ax):
    loc = mdates.AutoDateLocator()
    ax.xaxis.set_major_locator(loc)
    ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(loc))

def _plot_series_for_site(sitio_i):
    # Sub_B1
    s_sub = (subs.loc[subs["sitio"] == sitio_i, ["fecha", "Sub_B1"]]
                  .dropna()
                  .sort_values("fecha"))
    if not s_sub.empty:
        fig = plt.figure(figsize=(7.4, 3.2))
        ax = plt.gca()
        ax.plot(s_sub["fecha"], s_sub["Sub_B1"], marker="o", linewidth=2, color=EM_BLUE)
        ax.set_title(f"Sub_B1 (fisicoquímico) — {sitio_i}", fontsize=12, pad=8, color="#222")
        ax.set_xlabel("Fecha"); ax.set_ylabel("Sub_B1 (0–100)")
        ax.set_ylim(0, 110)
        ax.grid(True, alpha=0.3)
        _fmt_dates(ax)
        plt.tight_layout()
        plt.show()

    # E. coli
    s_ec = datos.loc[datos["sitio"] == sitio_i, ["fecha", "Ecoli_UFC100ml"]].copy()
    s_ec["Ecoli_UFC100ml"] = pd.to_numeric(s_ec["Ecoli_UFC100ml"], errors="coerce")
    # En escala log, valores <=0 no se pueden graficar
    s_ec = s_ec.dropna().query("Ecoli_UFC100ml > 0").sort_values("fecha")

    if not s_ec.empty:
        fig = plt.figure(figsize=(7.4, 3.2))
        ax = plt.gca()
        ax.plot(s_ec["fecha"], s_ec["Ecoli_UFC100ml"], marker="o", linewidth=2, color=EM_GREEN, label="E. coli")
        ax.set_yscale("log")
        # líneas guía desde configuración
        ax.axhline(EC_OBJ,  linestyle="--", linewidth=1.2, color="#666")
        ax.axhline(EC_CRIT, linestyle="--", linewidth=1.2, color="#666")
        # etiquetas guía (pegadas al borde izquierdo del eje de fechas)
        if not s_ec["fecha"].empty:
            x0 = s_ec["fecha"].min()
            ax.text(x0, EC_OBJ*1.05,  f"{int(EC_OBJ)}",  fontsize=9, color="#444")
            ax.text(x0, EC_CRIT*1.05, f"{int(EC_CRIT)}", fontsize=9, color="#444")

        ax.set_title(f"E. coli (UFC/100 mL) — {sitio_i}", fontsize=12, pad=8, color="#222")
        ax.set_xlabel("Fecha"); ax.set_ylabel("UFC/100 mL (log)")
        ax.grid(True, which="both", alpha=0.3)
        _fmt_dates(ax)
        plt.tight_layout()
        plt.show()

# ---------- Graficar todos los sitios con series ----------
for sitio_i in multi_series_sites:
    _plot_series_for_site(sitio_i)
```

### 6.2 — Comparación entre sitios

```python
# @title
# =============================================
# 6.2 — Comparación entre sitios (todas las SERIES) — UI EcoMonitor
#   - Por serie: Sub_B1 por sitio (barras 0–100)
#   - Resumen final (más pequeño): barras apiladas por serie
#       * Altura = Sub_B1 medio (0–100)
#       * Segmentos = distribución de categorías en esa serie
# =============================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import HTML, display

# ---- Ensanchar área de salida (sin scroll) ----
display(HTML("""
<style>
.output_wrapper, .output, .output_area {height:auto !important; max-height:none !important;}
.output_scroll {overflow-y: visible !important; height:auto !important; max-height:none !important; box-shadow:none !important;}
</style>
"""))

# ---- Estilo EcoMonitor ----
EM_BLUE  = globals().get("EM_BLUE",  "#135566")
EM_GREEN = globals().get("EM_GREEN", "#7BC043")
EM_LIGHT = globals().get("EM_LIGHT", "#F2F7F4")
STYLE_BOX = globals().get(
    "STYLE_BOX",
    f"background:{EM_LIGHT}; border-left:4px solid {EM_GREEN}; padding:10px 12px; border-radius:10px;"
)
def _card(html): display(HTML(f"<div style='{STYLE_BOX}; margin:12px 0 16px 0'>{html}</div>"))

# ---- Paletas (desde 6.0; fallback si faltan) ----
CAT_COLORS = globals().get("CAT_COLORS", {
    "Apta": "#2E7D32",
    "Aceptable": "#7CB342",
    "Condicional": "#F9A825",
    "Mala": "#F4511E",
    "No apta": "#C62828",
    "Indeterminada": "#9E9E9E",
})
def _cat_color(c): return CAT_COLORS.get(str(c), "#9E9E9E")

# ---- Chequeos mínimos ----
if 'subindices' not in globals():
    raise RuntimeError("No se encontró 'subindices'. Corré Sección 4.")
if 'df_datos' not in globals():
    raise RuntimeError("No se encontró 'df_datos'. Corré Sección 1.")
if 'df_clasificacion' not in globals():
    raise RuntimeError("No se encontró 'df_clasificacion'. Corré Sección 5.")

subs  = subindices.copy()
datos = df_datos.copy()
clf   = df_clasificacion.copy()

# Tipos
for d in (subs, datos, clf):
    if 'fecha' in d.columns:
        d['fecha'] = pd.to_datetime(d['fecha'], errors='coerce')

# ---- Resolver columna 'serie' con tolerancia (case-insensitive) ----
def _find_col_case_insensitive(df, name):
    low = {c.lower(): c for c in df.columns}
    return low.get(name.lower())

serie_col_datos = _find_col_case_insensitive(datos, "serie")
serie_col_subs  = _find_col_case_insensitive(subs,  "serie")
serie_col_clf   = _find_col_case_insensitive(clf,   "serie")

# Si ya viene en subindices/clf, normalizamos nombre; si no, intentamos unir desde df_datos
if serie_col_subs is None or serie_col_clf is None:
    # claves disponibles para unir
    key_pool = ['id','sitio','fecha']
    key_cols = [k for k in key_pool if k in subs.columns and k in datos.columns]
    if serie_col_datos is not None and key_cols:
        m_ser = datos[key_cols + [serie_col_datos]].drop_duplicates()
        m_ser = m_ser.rename(columns={serie_col_datos: 'serie'})
        if serie_col_subs is None:
            subs = subs.merge(m_ser, on=key_cols, how='left')
        if serie_col_clf is None:
            clf  = clf.merge(m_ser,  on=key_cols, how='left')
    else:
        # fallback: no hay 'serie' en datos o no hay claves -> crear 'General'
        if serie_col_subs is None:
            subs['serie'] = "General"
        else:
            subs = subs.rename(columns={serie_col_subs: 'serie'})
        if serie_col_clf is None:
            clf['serie'] = "General"
        else:
            clf = clf.rename(columns={serie_col_clf: 'serie'})
else:
    # normalizar nombre en ambos si ya existían
    if serie_col_subs != 'serie': subs = subs.rename(columns={serie_col_subs: 'serie'})
    if serie_col_clf  != 'serie': clf  = clf.rename(columns={serie_col_clf:  'serie'})

# Asegurar existencia y tipo string
if 'serie' not in subs.columns: subs['serie'] = "General"
if 'serie' not in clf.columns:  clf['serie']  = "General"
subs['serie'] = subs['serie'].astype(str).fillna("General")
clf['serie']  = clf['serie'].astype(str).fillna("General")

# ---- Orden de series por primera aparición temporal (o alfabético si no hay fecha) ----
if 'fecha' in subs.columns and subs['fecha'].notna().any():
    first_dt = subs.groupby('serie')['fecha'].min().sort_values()
    series_order = first_dt.index.tolist()
else:
    series_order = sorted(subs['serie'].unique().tolist())

# ---------- Tarjeta resumen ----------
n_series = len(series_order)
n_sites_total = subs['sitio'].nunique() if 'sitio' in subs.columns else 0
_card(f"""
  <div style="color:{EM_BLUE}; font-weight:800; font-size:20px; margin:0 0 6px 0;">
    6.2 — Comparación entre sitios y SERIES
  </div>
  <div style="border-top:2px solid {EM_BLUE}; margin:6px 0 10px 0; opacity:0.06;"></div>
  <div style="display:grid; grid-template-columns: 260px 1fr; gap:8px 16px; align-items:start; font-size:15px;">
    <div>Series detectadas</div><div><b>{n_series}</b></div>
    <div>Sitios totales</div><div><b>{n_sites_total}</b></div>
  </div>
  <div style="font-size:13px; color:#333; margin-top:8px;">
    Por cada <b>serie</b> se muestra <b>Sub_B1 por sitio</b> (0–100). Al final, un gráfico de <b>barras apiladas</b> por serie:
    <b>altura</b> = <b>Sub_B1 medio</b> (0–100) y <b>colores</b> = distribución de categorías.
  </div>
""")

# ---------- Gráficos por SERIE: SOLO Sub_B1 por sitio ----------
compact = (n_series > 3)

for serie in series_order:
    tmp = subs.loc[subs['serie'] == serie, ['sitio', 'Sub_B1']].copy()
    tmp['Sub_B1'] = pd.to_numeric(tmp['Sub_B1'], errors='coerce')
    tmp = tmp.dropna(subset=['Sub_B1'])
    if tmp.empty:
        continue

    per_site = tmp.groupby('sitio', as_index=False)['Sub_B1'].mean()
    per_site = per_site.sort_values('Sub_B1', ascending=False)

    # Tamaño (compacto si muchas series)
    width = 7.8 if not compact else 6.6
    height = 3.2

    plt.figure(figsize=(width, height))
    plt.bar(per_site['sitio'].astype(str), per_site['Sub_B1'].astype(float), color=EM_BLUE)
    plt.title(f"Serie: {serie} — Sub_B1 por sitio", fontsize=11, color=EM_BLUE, pad=8)
    plt.xlabel("Sitio", fontsize=10); plt.ylabel("Sub_B1 (0–100)", fontsize=10)
    plt.ylim(0, 110)
    plt.xticks(rotation=45, ha="right", fontsize=9 if not compact else 8)
    plt.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.show()

# ---------- Barras apiladas por SERIE (más pequeño, como los otros)
cat_order = ["Apta","Aceptable","Condicional","Mala","No apta","Indeterminada"]
rows = []

for serie in series_order:
    s = subs.loc[subs['serie'] == serie, ['Sub_B1']].copy()
    s['Sub_B1'] = pd.to_numeric(s['Sub_B1'], errors='coerce').dropna()
    if s.empty:
        continue
    mean_b1 = float(s['Sub_B1'].mean())

    cc = clf.loc[clf['serie'] == serie, ['sitio','fecha','Clasificacion_final']].dropna(subset=['sitio'])
    if not cc.empty:
        cc['fecha'] = pd.to_datetime(cc['fecha'], errors='coerce')
        cc = cc.sort_values(['sitio','fecha']).groupby('sitio', as_index=False).tail(1)
        total = cc['sitio'].nunique()
        counts = cc['Clasificacion_final'].value_counts()
        shares = {cat: (counts.get(cat, 0)/total if total>0 else 0.0) for cat in cat_order}
    else:
        shares = {cat: 0.0 for cat in cat_order}

    rows.append({
        "serie": serie,
        "mean_b1": mean_b1,
        **{f"seg_{cat}": shares[cat]*mean_b1 for cat in cat_order}
    })

if rows:
    dfp = pd.DataFrame(rows).set_index('serie')
    dfp = dfp.reindex([s for s in series_order if s in dfp.index])

    width = 6.9
    height = 4.0  # más pequeño
    plt.figure(figsize=(width, height))

    bottoms = np.zeros(len(dfp))
    xlabels = dfp.index.astype(str).tolist()

    for cat in cat_order:
        col = f"seg_{cat}"
        vals = dfp[col].values if col in dfp.columns else np.zeros(len(dfp))
        plt.bar(xlabels, vals, bottom=bottoms, color=_cat_color(cat), edgecolor="white", linewidth=0.4, label=cat)
        bottoms += vals

    # Etiquetas con Sub_B1 medio
    for i, mean_b1 in enumerate(dfp["mean_b1"].values):
        plt.text(i, mean_b1 + 1.0, f"{mean_b1:.0f}", ha="center", va="bottom", fontsize=8.5, color="#333")

    plt.ylim(0, 110)
    plt.ylabel("Sub_B1 medio (0–100)", fontsize=10)
    plt.title("Series — Sub_B1 medio y composición de categorías", fontsize=11, color=EM_BLUE, pad=8)
    plt.xticks(rotation=0, fontsize=9)
    plt.grid(axis="y", alpha=0.25)
    plt.legend(ncol=min(6, len(cat_order)), fontsize=8, frameon=False,
               loc="upper center", bbox_to_anchor=(0.5, -0.12))
    plt.tight_layout(rect=[0, 0.04, 1, 1])
    plt.show()
else:
    _card("ℹ️ No hay datos suficientes para el gráfico apilado por serie (Sub_B1 medio + categorías).")
```

### 6.4 — Heatmap de drivers (parámetros más críticos)

```python
# @title
# =============================================
# 6.4 — Variables más críticas (Bloque 1) — por serie
#   - Un solo encabezado general
#   - Un heatmap por serie (columna `serie` del dataset)
#   - Solo variables activas del Sub_B1 (Sección 3)
#   - Colores: rojo (crítico) → amarillo → verde (cumple)
# =============================================
import numpy as np, pandas as pd, warnings
from IPython.display import display, HTML

warnings.filterwarnings("ignore", category=FutureWarning)

# ---- Estilo EcoMonitor ----
EM_BLUE  = globals().get("EM_BLUE",  "#1F3B64")
EM_GREEN = globals().get("EM_GREEN", "#7BC043")
EM_LIGHT = globals().get("EM_LIGHT", "#F2F7F4")
STYLE_BOX = globals().get("STYLE_BOX",
    f"background:{EM_LIGHT}; border-left:4px solid {EM_GREEN}; padding:12px 14px; border-radius:12px;"
)
def _card(html): display(HTML(f"<div style='{STYLE_BOX}; margin:12px 0 16px 0'>{html}</div>"))

# ---- Chequeos / datos ----
if 'scores_vars' not in globals() or scores_vars is None or scores_vars.empty:
    raise RuntimeError("No se encontró 'scores_vars'. Corré la Sección 4 después de definir la selección en Sección 3.")

sv = scores_vars.copy()

# Asegurar columnas clave
for col in ["id","sitio","fecha","parametro","score"]:
    if col not in sv.columns:
        raise RuntimeError(f"'scores_vars' no contiene la columna requerida: {col}")

# Incorporar 'serie' desde scores_vars o, si falta, desde df_datos
if "serie" not in sv.columns or sv["serie"].isna().all():
    if 'df_datos' in globals() and {"id","sitio","fecha","serie"}.issubset(df_datos.columns):
        sv = sv.merge(
            df_datos[["id","sitio","fecha","serie"]],
            on=["id","sitio","fecha"], how="left"
        )
    else:
        # Fallback: una sola serie "General"
        sv["serie"] = "General"

# Normalizar tipos mínimos
sv["fecha"] = pd.to_datetime(sv["fecha"], errors="coerce")

# ---- Filtrar SOLO variables activas del Bloque 1 (Sub_B1) ----
active_vars = None
cfg = None
if 'config_b1b2' in globals() and config_b1b2 is not None:
    cfg = config_b1b2.copy()
elif 'ref_rec' in globals() and ref_rec is not None:
    cfg = ref_rec.copy()

if cfg is not None:
    cfg = cfg.copy()
    cfg.columns = [c.strip().lower() for c in cfg.columns]
    for col in ["bloque","sentido","peso","parametro"]:
        if col not in cfg.columns: cfg[col] = np.nan

    m_b1 = cfg["bloque"].astype(str).str.upper().eq("B1")
    m_on = cfg["sentido"].astype(str).str.strip().ne("")  # activas (tienen sentido)
    m_w  = pd.to_numeric(cfg["peso"], errors="coerce").fillna(0) > 0  # con peso > 0
    active_vars = set(cfg.loc[m_b1 & m_on & m_w, "parametro"].astype(str).str.strip())

elif 'SEL' in globals() and isinstance(SEL, dict):
    # Fallback: selección de la Sección 2/3 si existe
    active_vars = set(SEL.get("B1", []))

# Aplicar filtro de variables activas si hay lista
if active_vars:
    sv = sv[sv["parametro"].astype(str).isin(active_vars)].copy()

if sv.empty:
    _card("ℹ️ No hay puntuaciones de variables activas del Bloque 1 para mostrar. Volvé a correr la Sección 4 tras ajustar la Sección 3.")
else:
    # Universo de variables y Top-N global
    all_vars = sorted(sv["parametro"].astype(str).unique().tolist())
    TOPN = min(12, len(all_vars)) if len(all_vars) > 0 else 0

    # ---- Encabezado general (una sola vez) ----
    _card(f"""
      <div style="color:{EM_BLUE}; font-weight:800; font-size:20px; margin:0 0 6px 0;">
        6.4 — Variables más críticas (Bloque 1)
      </div>
      <div style="font-size:14px; color:#333; margin:4px 0 0 0;">
        Variables mostradas: <b>Top {TOPN}</b> (según menor puntuación media por serie).
        Colores: <b style="color:#C62828">rojo</b> (crítico) → <b style="color:#F9A825">amarillo</b> → <b style="color:#2E7D32">verde</b> (cumple).
      </div>
    """)

    # ---- Helpers de color ----
    def _rgb_for_score(v):
        try: v = float(v)
        except: return (240,240,240)
        if np.isnan(v): return (240,240,240)
        v = min(100.0, max(0.0, v))
        if v <= 50:
            t = v/50.0
            r = int(198*(1-t) + 249*t); g = int(40*(1-t) + 168*t); b = int(40*(1-t) + 37*t)
        else:
            t = (v-50)/50.0
            r = int(249*(1-t) + 46*t);  g = int(168*(1-t) + 125*t); b = int(37*(1-t) + 50*t)
        return (r,g,b)

    def _cell_style(v):
        r,g,b = _rgb_for_score(v)
        # contraste de texto por luminancia
        Y = (0.2126*r + 0.7152*g + 0.0722*b) / 255.0
        txt = "#000" if Y > 0.6 else "#fff"
        return f"background-color: rgb({r},{g},{b}); color:{txt}; font-weight:700;"

    # ---- Un heatmap por serie ----
    series_vals = [s for s in sv["serie"].astype(str).fillna("—").unique().tolist()]
    series_vals = sorted(series_vals)

    for serie in series_vals:
        sv_c = sv.loc[sv["serie"].astype(str) == str(serie)].copy()
        if sv_c.empty:
            continue

        # Pivot: promedio por sitio–parámetro (cuanto más bajo el score, más crítico)
        piv = (sv_c.groupby(["sitio","parametro"])["score"]
                    .mean()                 # sin FutureWarning
                    .unstack("parametro"))

        if piv is None or piv.empty:
            continue

        # Ordenar columnas por score medio (peores primero) y tomar Top-N global
        order_cols = piv.mean(axis=0, skipna=True).sort_values(ascending=True).index.tolist()
        n_use = min(TOPN, len(order_cols)) if TOPN > 0 else len(order_cols)
        piv_top = piv[order_cols[:n_use]]

        # Título por serie
        display(HTML(
            f"<div style='font-weight:800; color:{EM_BLUE}; font-size:16px; margin:22px 0 6px 0;'>"
            f"Serie: <span style='color:#111'>{serie}</span></div>"
        ))

        # Encabezados de ejes
        piv_top.index.name = "Sitio"
        piv_top.columns.name = "Variable"

        # Render del heatmap (pandas Styler) con etiquetas de sitios visibles
        styled = (
            piv_top.style
            .format("{:.0f}")
            .set_properties(**{"font-size":"12px","white-space":"nowrap"})
            .set_table_styles([
                {"selector": "th.row_heading", "props": [("font-weight","700"), ("text-align","left")]},
                {"selector": "th.col_heading", "props": [("font-weight","700"), ("text-align","center")]},
            ])
            .map(_cell_style)
        )
        display(styled)
```

```python
# @title
# =============================================
# 6.5 — Mapa interactivo (por serie) — UI EcoMonitor
# - Filtro por SERIE (columna 'serie' / 'Serie' / 'SERIE'); + opción "General (promedio)"
# - Dato: Sub_B1 (promedio), Estado_B2, Clasificación final
# - Tiles: OSM estándar, Esri satélite, CartoDB claro
# - Descarga HTML a PANTALLA COMPLETA (sin recortes)
# =============================================
import numpy as np
import pandas as pd
import ipywidgets as w
from IPython.display import display, HTML, clear_output

# ---- Estilo EcoMonitor (para botón) ----
EM_BLUE  = globals().get("EM_BLUE",  "#135566")
EM_GREEN = globals().get("EM_GREEN", "#7BC043")
EM_LIGHT = globals().get("EM_LIGHT", "#F2F7F4")
STYLE_BOX = globals().get(
    "STYLE_BOX",
    f"background:{EM_LIGHT}; border-left:4px solid {EM_GREEN}; padding:10px 12px; border-radius:10px;"
)
display(HTML(f"""
<style>
.em-btn.widget-button {{
  border-radius: 12px !important;
  font-weight: 700 !important;
  border: none !important;
  width: 240px !important; height: 40px !important;
  margin-top: 8px !important;
}}
.em-btn-blue.widget-button  {{ background: {EM_BLUE} !important; color:#fff !important; }}
</style>
"""))

def _card(html): display(HTML(f"<div style='{STYLE_BOX}; margin:12px 0 10px 0'>{html}</div>"))

# ---- Chequeos ----
if 'clf' not in globals():
    raise RuntimeError("No se encontró 'clf' (df_clasificacion). Corré 6.0.")
if 'sitios' not in globals():
    raise RuntimeError("No se encontró 'sitios' (df_sitios). Corré la Sección 1 y luego 6.0.")

# Paletas / helpers
_cat_color = globals().get("_cat_color", lambda x:"#9E9E9E")
CAT_COLORS = globals().get("CAT_COLORS", {
    "Apta": "#2E7D32", "Aceptable": "#7CB342", "Condicional": "#F9A825",
    "Mala": "#F4511E", "No apta": "#C62828", "Indeterminada": "#9E9E9E",
})
SAN_COLORS = globals().get("SAN_COLORS", {
    "Seguro": "#2E7D32", "Comprometido": "#F9A825", "No apto": "#C62828", "Sin dato": "#9E9E9E",
})
B1_COLORS = {"Buena": "#2E7D32", "Moderada": "#F9A825", "Mala": "#C62828", "Sin dato": "#9E9E9E"}

def _clase_b1(v):
    if pd.isna(v): return "Sin dato"
    v=float(v)
    if v>=80: return "Buena"
    if v>=50: return "Moderada"
    return "Mala"

def _latest_by_site(df, date_col="fecha"):
    d=df.copy()
    d["fecha"]=pd.to_datetime(d["fecha"], errors="coerce")
    return d.sort_values(["sitio",date_col]).groupby("sitio",as_index=False).tail(1)

# ---- Normalización de datos ----
dfc = clf.copy()
if 'fecha' in dfc.columns:
    dfc['fecha']=pd.to_datetime(dfc['fecha'], errors='coerce')
dsit = sitios.copy()

# Detectar columna SERIE (case-insensitive)
serie_col = None
for c in dfc.columns:
    if str(c).strip().lower()=="serie":
        serie_col = c
        break
if serie_col is None:
    dfc["_serie_label"] = "Única"
else:
    dfc["_serie_label"] = dfc[serie_col].astype(str).str.strip().replace({"": np.nan}).fillna("Sin serie")

# Lista de series preservando orden
seen=set(); series_labels=[]
for s in dfc["_serie_label"]:
    if s not in seen:
        series_labels.append(s); seen.add(s)

# ---- Widgets ----
dd_serie = w.Dropdown(
    options=[("General (promedio)", "__general__")] + [(s,s) for s in series_labels],
    value=(series_labels[-1] if series_labels else "__general__"),
    description="Serie:", layout=w.Layout(width="320px")
)
dd_dato = w.Dropdown(
    options=["Sub_B1 (promedio)", "Estado_B2", "Clasificación final"],
    value="Sub_B1 (promedio)", description="Dato:", layout=w.Layout(width="320px")
)

TILE_PRESETS = {
    "OSM estándar": {"builtin":"OpenStreetMap"},
    "Esri WorldImagery (satélite)": {
        "url":"https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}",
        "attr":"Tiles © Esri — Source: Esri, USGS, NOAA, etc."
    },
    "CartoDB (claro)": {"builtin":"CartoDB positron"},
}
dd_tiles = w.Dropdown(
    options=list(TILE_PRESETS.keys()), value="OSM estándar",
    description="Mapa:", layout=w.Layout(width="320px")
)
dd_tiles.style.description_width = "70px"  # ancho del label "Mapa:"

btn_desc = w.Button(description="📥 Descargar mapa (HTML)")
btn_desc.add_class("em-btn"); btn_desc.add_class("em-btn-blue")
# Alinear con el inicio de “Mapa”
btn_desc.layout = w.Layout(margin="0 0 0 76px")

out_map = w.Output()

# ---- Preparación de dataset según selección ----
def _prepare_mdf(sel_series, modo):
    if modo == "Sub_B1 (promedio)":
        if sel_series == "__general__":
            base = dfc.groupby("sitio", as_index=False)["Sub_B1"].mean()
            base["fecha"] = _latest_by_site(dfc)["fecha"]
            serie_txt = "General (promedio)"
        else:
            base = (dfc[dfc["_serie_label"]==sel_series]
                    .groupby("sitio", as_index=False)["Sub_B1"].mean())
            fref = _latest_by_site(dfc[dfc["_serie_label"]==sel_series])[["sitio","fecha"]]
            base = base.merge(fref, on="sitio", how="left")
            serie_txt = sel_series
        base["Clase_B1"] = base["Sub_B1"].apply(_clase_b1)
        mdf = pd.merge(dsit, base, on="sitio", how="left")
        ttl = "6.5 — Mapa (Sub_B1 por sitio)"
        legend = " · ".join([f"<span style='display:inline-block;width:12px;height:12px;border-radius:2px;background:{c};margin-right:6px;'></span>{k}"
                             for k,c in [("Buena",B1_COLORS["Buena"]),("Moderada",B1_COLORS["Moderada"]),("Mala",B1_COLORS["Mala"])]])
        def _color_row(r): return B1_COLORS.get(str(r.get("Clase_B1","Sin dato")), "#9E9E9E")
        def _popup(r):
            sitio=str(r.get("sitio","")); desc=(r.get("descripcion","") or "")
            fecha_txt = pd.to_datetime(r.get("fecha")).date().isoformat() if pd.notna(r.get("fecha")) else "—"
            v = r.get("Sub_B1", np.nan); vtxt = "—" if pd.isna(v) else f"{float(v):.1f}"
            cl = str(r.get("Clase_B1","Sin dato")); col=_color_row(r)
            return (f"<div style='min-width:220px'>"
                    f"<div style='font-weight:800; color:{EM_BLUE}; font-size:14px;'>{sitio}</div>"
                    f"<div style='font-size:12px; color:#333;'>{desc}</div>"
                    f"<div style='margin-top:6px; font-size:13px;'>"
                    f"<b>Serie:</b> {serie_txt}<br>"
                    f"<b>Sub_B1:</b> {vtxt} "
                    f"(<span style='background:{col}; color:#fff; padding:2px 6px; border-radius:999px; font-weight:700;'>{cl}</span>)"
                    f"</div></div>")
        return mdf, ttl, legend, _color_row, _popup, serie_txt

    elif modo == "Estado_B2":
        dsel = _latest_by_site(dfc if sel_series=="__general__" else dfc[dfc["_serie_label"]==sel_series])
        mdf = pd.merge(dsit, dsel[["sitio","fecha","Estado_B2"]], on="sitio", how="left")
        serie_txt = "Última disponible" if sel_series=="__general__" else sel_series
        ttl = "6.5 — Mapa (Estado sanitario por sitio)"
        order = [("Seguro",SAN_COLORS["Seguro"]),("Comprometido",SAN_COLORS["Comprometido"]),
                 ("No apto",SAN_COLORS["No apto"]),("Sin dato",SAN_COLORS["Sin dato"])]
        legend = " · ".join([f"<span style='display:inline-block;width:12px;height:12px;border-radius:2px;background:{c};margin-right:6px;'></span>{k}" for k,c in order])
        def _color_row(r): return SAN_COLORS.get(str(r.get("Estado_B2","Sin dato")), "#9E9E9E")
        def _popup(r):
            sitio=str(r.get("sitio","")); desc=(r.get("descripcion","") or "")
            fecha_txt = pd.to_datetime(r.get("fecha")).date().isoformat() if pd.notna(r.get("fecha")) else "—"
            st=str(r.get("Estado_B2","Sin dato")); col=_color_row(r)
            return (f"<div style='min-width:220px'>"
                    f"<div style='font-weight:800; color:{EM_BLUE}; font-size:14px;'>{sitio}</div>"
                    f"<div style='font-size:12px; color:#333;'>{desc}</div>"
                    f"<div style='margin-top:6px; font-size:13px;'>"
                    f"<b>Serie:</b> {serie_txt}<br>"
                    f"<b>Sanitario:</b> <span style='background:{col}; color:#fff; padding:2px 6px; border-radius:999px; font-weight:700;'>{st}</span>"
                    f"</div></div>")
        return mdf, ttl, legend, _color_row, _popup, serie_txt

    else:  # Clasificación final
        dsel = _latest_by_site(dfc if sel_series=="__general__" else dfc[dfc["_serie_label"]==sel_series])
        mdf = pd.merge(dsit, dsel[["sitio","fecha","Clasificacion_final"]], on="sitio", how="left")
        serie_txt = "Última disponible" if sel_series=="__general__" else sel_series
        ttl = "6.5 — Mapa (Clasificación final por sitio)"
        order = ["Apta","Aceptable","Condicional","Mala","No apta","Indeterminada"]
        legend = " · ".join([f"<span style='display:inline-block;width:12px;height:12px;border-radius:2px;background:{CAT_COLORS.get(k,'#9E9E9E')};margin-right:6px;'></span>{k}" for k in order])
        def _color_row(r): return _cat_color(str(r.get("Clasificacion_final","Indeterminada")))
        def _popup(r):
            sitio=str(r.get("sitio","")); desc=(r.get("descripcion","") or "")
            fecha_txt = pd.to_datetime(r.get("fecha")).date().isoformat() if pd.notna(r.get("fecha")) else "—"
            cat=str(r.get("Clasificacion_final","Indeterminada")); col=_color_row(r)
            return (f"<div style='min-width:220px'>"
                    f"<div style='font-weight:800; color:{EM_BLUE}; font-size:14px;'>{sitio}</div>"
                    f"<div style='font-size:12px; color:#333;'>{desc}</div>"
                    f"<div style='margin-top:6px; font-size:13px;'>"
                    f"<b>Serie:</b> {serie_txt}<br>"
                    f"<b>Clasificación:</b> <span style='color:{col}; font-weight:800'>{cat}</span>"
                    f"</div></div>")
        return mdf, ttl, legend, _color_row, _popup, serie_txt

# ---- Construcción del mapa (display) ----
def _make_map(mdf, color_fn, popup_fn, tile_key, height_css="520px"):
    import folium
    m = folium.Map(
        location=[float(mdf['latitud'].mean()), float(mdf['longitud'].mean())],
        tiles=None, zoom_start=11, control_scale=True,
        width="100%", height=height_css
    )
    preset = TILE_PRESETS[tile_key]
    if "builtin" in preset:
        folium.TileLayer(preset["builtin"], control=False).add_to(m)
    else:
        folium.TileLayer(tiles=preset["url"], attr=preset["attr"], control=False).add_to(m)

    coords=[]
    for _, r in mdf.iterrows():
        lat, lon = float(r["latitud"]), float(r["longitud"])
        col = color_fn(r)
        folium.CircleMarker(
            location=[lat,lon], radius=7, color=col, fill=True, fill_color=col, fill_opacity=0.9,
            popup=folium.Popup(popup_fn(r), max_width=320)
        ).add_to(m)
        coords.append((lat,lon))
    if coords:
        m.fit_bounds(coords, padding=(14,14))
    return m

# ---- Construcción del mapa (export FULL WINDOW) ----
def _make_map_export(mdf, color_fn, popup_fn, tile_key):
    import folium
    from branca.element import Element
    m = _make_map(mdf, color_fn, popup_fn, tile_key, height_css="100%")
    # CSS para ocupar toda la ventana en el HTML exportado
    css = """
    <style>
      html, body {height:100%; width:100%; margin:0; padding:0;}
      .folium-map {position:absolute; top:0; bottom:0; left:0; right:0;}
    </style>
    """
    m.get_root().html.add_child(Element(css))
    return m

# ---- Render principal ----
def _build_map(_=None):
    with out_map:
        clear_output()
        try:
            import folium
        except Exception:
            _card("🗺️ Necesitás <b>folium</b>. Instalá: <code>!pip install folium</code>")
            return

        sel_series  = dd_serie.value
        modo        = dd_dato.value
        mdf, ttl, legend, color_fn, popup_fn, serie_txt = _prepare_mdf(sel_series, modo)

        # Coords válidas
        mdf["latitud"]  = pd.to_numeric(mdf.get("latitud"),  errors="coerce")
        mdf["longitud"] = pd.to_numeric(mdf.get("longitud"), errors="coerce")
        mdf = mdf.dropna(subset=["latitud","longitud"])
        if mdf.empty:
            _card("🗺️ No hay coordenadas válidas para mostrar.")
            return

        fmap = _make_map(mdf, color_fn, popup_fn, dd_tiles.value, height_css="560px")  # compacto y sin blanco extra

        _card(f"""
          <div style="color:{EM_BLUE}; font-weight:800; font-size:20px; margin:0 0 6px 0;">{ttl}</div>
          <div style="border-top:2px solid {EM_BLUE}; margin:6px 0 8px 0; opacity:0.06;"></div>
          <div style="font-size:14px; color:#333; margin-bottom:4px;">
            Serie: <b>{'General (promedio)' if dd_serie.value=='__general__' else dd_serie.value}</b> · Dato: <b>{modo}</b>
          </div>
          <div style="font-size:13px; color:#333;">{legend}</div>
        """)
        display(fmap)

        # Guardar payload para export
        btn_desc._payload = dict(sel_series=sel_series, modo=modo, tile=dd_tiles.value)

        # Descargar
        def _save_map(_):
            path = "/content/mapa_interactivo.html"
            try:
                payload = getattr(btn_desc, "_payload", None) or {}
                sel_series = payload.get("sel_series", dd_serie.value)
                modo       = payload.get("modo", dd_dato.value)
                tile_key   = payload.get("tile", dd_tiles.value)

                mdf, *_rest = _prepare_mdf(sel_series, modo)
                mdf["latitud"]  = pd.to_numeric(mdf.get("latitud"),  errors="coerce")
                mdf["longitud"] = pd.to_numeric(mdf.get("longitud"), errors="coerce")
                mdf = mdf.dropna(subset=["latitud","longitud"])
                _, _, _, color_fn, popup_fn, _ = _prepare_mdf(sel_series, modo)

                m_export = _make_map_export(mdf, color_fn, popup_fn, tile_key)
                m_export.save(path)
                try:
                    from google.colab import files
                    files.download(path)
                except Exception:
                    _card(f"📄 Mapa exportado en <code>{path}</code>")
            except Exception as e:
                _card(f"❌ No se pudo exportar el mapa: {e}")

        try: btn_desc._click_handlers.callbacks.clear()
        except Exception: pass
        btn_desc.on_click(_save_map)

# UI
controls_left  = w.VBox([dd_serie, dd_dato], layout=w.Layout(width="340px"))
controls_right = w.VBox([dd_tiles, btn_desc], layout=w.Layout(width="340px"))
ui = w.HBox([controls_left, controls_right], layout=w.Layout(gap="16px", align_items="flex-start"))

_card(f"""
  <div style="color:{EM_BLUE}; font-weight:800; font-size:20px; margin:0 0 6px 0;">
    6.5 — Mapa interactivo (por serie)
  </div>
  <div style="border-top:2px solid {EM_BLUE}; margin:6px 0 8px 0; opacity:0.06;"></div>
  <div style="font-size:14px; color:#333;">Elegí la <b>serie</b> (o <b>General</b>), el dato y el estilo de mapa.</div>
""")
display(ui)
out_map.clear_output(wait=True); display(out_map)

# Inicial + listeners
_build_map()
for wg in (dd_serie, dd_dato, dd_tiles):
    wg.observe(_build_map, names="value")
```

```python
# @title
# =============================================
# 6.6 — Exportación de resultados (Excel + PNGs + Mapa + ZIP) — UI EcoMonitor
# Requiere: df_clasificacion (Sec. 5), subindices (Sec. 4), df_datos (Sec. 1)
# =============================================
import os, io, zipfile, datetime, re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML

# ---------- Estilo EcoMonitor ----------
EM_BLUE  = globals().get("EM_BLUE",  "#135566")
EM_GREEN = globals().get("EM_GREEN", "#7BC043")
EM_LIGHT = globals().get("EM_LIGHT", "#F2F7F4")
STYLE_BOX = globals().get(
    "STYLE_BOX",
    f"background:{EM_LIGHT}; border-left:4px solid {EM_GREEN}; padding:10px 12px; border-radius:10px;"
)
display(HTML(f"""
<style>
.em-btn.widget-button {{
  border-radius: 12px !important;
  font-weight: 700 !important;
  border: none !important;
  width: 260px !important; height: 40px !important;
  margin: 8px 8px 0 0 !important;
}}
.em-btn-blue.widget-button  {{ background: {EM_BLUE} !important; color:#fff !important; }}
.em-btn-green.widget-button {{ background: {EM_GREEN} !important; color:#fff !important; }}

.output_wrapper, .output, .output_area {{ height:auto !important; max-height:none !important; }}
.output_scroll {{ overflow-y: visible !important; height:auto !important; max-height:none !important; box-shadow:none !important; }}
</style>
"""))

def _card(html): display(HTML(f"<div style='{STYLE_BOX}; margin:12px 0 16px 0'>{html}</div>"))

# ---------- Chequeos de insumos mínimos ----------
missing = []
if 'df_clasificacion' not in globals(): missing.append("df_clasificacion (Sección 5)")
if 'subindices' not in globals():       missing.append("subindices (Sección 4)")
if 'df_datos' not in globals():         missing.append("df_datos (Sección 1)")
if missing:
    raise RuntimeError("Faltan insumos para exportar: " + ", ".join(missing))

clf   = df_clasificacion.copy()
subs  = subindices.copy()
datos = df_datos.copy()
sitios = globals().get('df_sitios', pd.DataFrame())
umbr   = globals().get('ref_rec',  pd.DataFrame())
scores = globals().get('scores_vars', pd.DataFrame())

for d in (clf, subs, datos):
    if 'fecha' in d.columns:
        d['fecha'] = pd.to_datetime(d['fecha'], errors='coerce')

# ---------- Resolver SERIE de forma robusta ----------
def _find_col_case_insensitive(df, name):
    low = {c.lower(): c for c in df.columns}
    return low.get(name.lower())

serie_datos = _find_col_case_insensitive(datos, "serie")
serie_subs  = _find_col_case_insensitive(subs,  "serie")
serie_clf   = _find_col_case_insensitive(clf,   "serie")

# Si subs/clf no tienen 'serie', intentamos unir desde df_datos por claves
def _ensure_serie(subs, clf, datos):
    serie_d = _find_col_case_insensitive(datos, "serie")
    # normalizar si ya existen
    if _find_col_case_insensitive(subs, "serie"): subs = subs.rename(columns={_find_col_case_insensitive(subs,"serie"): "serie"})
    if _find_col_case_insensitive(clf,  "serie"): clf  = clf.rename(columns={_find_col_case_insensitive(clf,"serie"):  "serie"})
    if "serie" not in subs.columns or "serie" not in clf.columns:
        keys = [k for k in ['id','sitio','fecha'] if k in subs.columns and k in datos.columns]
        if serie_d is not None and keys:
            m = datos[keys + [serie_d]].drop_duplicates().rename(columns={serie_d: "serie"})
            if "serie" not in subs.columns: subs = subs.merge(m, on=keys, how="left")
            if "serie" not in clf.columns:  clf  = clf.merge(m, on=keys,  how="left")
        else:
            if "serie" not in subs.columns: subs["serie"] = "General"
            if "serie" not in clf.columns:  clf["serie"]  = "General"
    subs["serie"] = subs["serie"].astype(str).fillna("General")
    clf["serie"]  = clf["serie"].astype(str).fillna("General")
    return subs, clf

subs, clf = _ensure_serie(subs, clf, datos)

# Orden de series
if 'fecha' in subs.columns and subs['fecha'].notna().any():
    series_order = subs.groupby('serie')['fecha'].min().sort_values().index.tolist()
else:
    series_order = sorted(subs['serie'].unique().tolist())

# ---------- Paletas ----------
CAT_COLORS = globals().get("CAT_COLORS", {
    "Apta": "#2E7D32","Aceptable": "#7CB342","Condicional": "#F9A825",
    "Mala": "#F4511E","No apta": "#C62828","Indeterminada": "#9E9E9E",
})
def _cat_color(c): return CAT_COLORS.get(str(c), "#9E9E9E")

# Carpeta de salida
out_dir = "/content/EcoMonitor_WQI_exports"
os.makedirs(out_dir, exist_ok=True)

_card(f"""
  <div style="color:{EM_BLUE}; font-weight:800; font-size:20px; margin:0 0 6px 0;">
    6.6 — Exportación de resultados
  </div>
  <div style="border-top:2px solid {EM_BLUE}; margin:6px 0 10px 0; opacity:0.06;"></div>
  <div style="font-size:15px; color:#333;">
    Se generará un paquete con:<br>
    • Excel consolidado (tablas clave).<br>
    • Gráficos PNG por <b>serie</b> y series temporales.<br>
    • Mapa interactivo (HTML) con la última medición por sitio.<br>
    • Archivo ZIP final con todo.
  </div>
""")

# ---------------------------------------------
# 1) Excel consolidado
# ---------------------------------------------
excel_path = os.path.join(out_dir, "EcoMonitor_WQI_resultados.xlsx")
try:
    import openpyxl  # noqa
except Exception:
    raise RuntimeError("Falta 'openpyxl'. Instalalo con: !pip install openpyxl")

with pd.ExcelWriter(excel_path, engine="openpyxl", mode="w") as writer:
    clf.to_excel(writer, sheet_name="clasificacion", index=False)
    subs.to_excel(writer, sheet_name="subindices", index=False)
    if isinstance(scores, pd.DataFrame) and not scores.empty:
        scores.to_excel(writer, sheet_name="scores_vars", index=False)
    if isinstance(umbr, pd.DataFrame) and not umbr.empty:
        u = umbr.copy()
        u.columns = [str(c).strip().lower() for c in u.columns]
        exp_cols = ["parametro","unidad","sentido","objetivo","objetivo_max","peso","fuente","nota"]
        for c in exp_cols:
            if c not in u.columns: u[c] = np.nan
        u[exp_cols].to_excel(writer, sheet_name="umbrales", index=False)
    datos.to_excel(writer, sheet_name="datos", index=False)
    if isinstance(sitios, pd.DataFrame) and not sitios.empty:
        sitios.to_excel(writer, sheet_name="sitios", index=False)

_card(f"📄 <b>Excel</b> creado: <code>{excel_path}</code>")

# ---------------------------------------------
# 2) Gráficos (PNGs) — coherentes con 6.2 (SERIES)
# ---------------------------------------------
def _slug(s):
    return re.sub(r'[^0-9A-Za-z_-]+', '_', str(s)).strip('_') or "serie"

png_paths = []

# 2.1 Barras Sub_B1 por sitio (una imagen por serie)
for serie in series_order:
    s = subs.loc[subs['serie'] == serie, ['sitio','Sub_B1']].copy()
    s['Sub_B1'] = pd.to_numeric(s['Sub_B1'], errors='coerce')
    s = s.dropna()
    if s.empty:
        continue
    per_site = s.groupby('sitio', as_index=False)['Sub_B1'].mean().sort_values('Sub_B1', ascending=False)

    plt.figure(figsize=(7.2, 3.2))
    plt.bar(per_site['sitio'].astype(str), per_site['Sub_B1'].astype(float), color=EM_BLUE)
    plt.title(f"Serie: {serie} — Sub_B1 por sitio", fontsize=11, color=EM_BLUE, pad=8)
    plt.xlabel("Sitio", fontsize=10); plt.ylabel("Sub_B1 (0–100)", fontsize=10)
    plt.ylim(0, 110); plt.xticks(rotation=45, ha="right", fontsize=9)
    plt.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    p = os.path.join(out_dir, f"serie_{_slug(serie)}__SubB1_por_sitio.png")
    plt.savefig(p, bbox_inches="tight", dpi=150); plt.close()
    png_paths.append(p)

# 2.2 Resumen apilado por serie (altura = Sub_B1 medio; segmentos = categorías)
cat_order = ["Apta","Aceptable","Condicional","Mala","No apta","Indeterminada"]
rows = []
for serie in series_order:
    s = subs.loc[subs['serie']==serie, ['Sub_B1']].copy()
    s['Sub_B1'] = pd.to_numeric(s['Sub_B1'], errors='coerce')
    s = s.dropna()
    if s.empty:
        continue
    mean_b1 = float(s['Sub_B1'].mean())

    cc = clf.loc[clf['serie']==serie, ['sitio','fecha','Clasificacion_final']].dropna(subset=['sitio'])
    if not cc.empty:
        cc['fecha'] = pd.to_datetime(cc['fecha'], errors='coerce')
        cc = cc.sort_values(['sitio','fecha']).groupby('sitio', as_index=False).tail(1)
        total = cc['sitio'].nunique()
        counts = cc['Clasificacion_final'].value_counts()
        shares = {k: (counts.get(k,0)/total if total>0 else 0.0) for k in cat_order}
    else:
        shares = {k:0.0 for k in cat_order}

    rows.append({
        "serie": serie,
        "mean_b1": mean_b1,
        **{f"seg_{k}": shares[k]*mean_b1 for k in cat_order}
    })

if rows:
    dfp = pd.DataFrame(rows).set_index('serie').reindex([s for s in series_order if s in pd.DataFrame(rows)['serie'].tolist()])
    plt.figure(figsize=(6.9, 4.0))
    bottoms = np.zeros(len(dfp))
    xlabels = dfp.index.astype(str).tolist()
    for cat in cat_order:
        vals = dfp[f"seg_{cat}"].values if f"seg_{cat}" in dfp.columns else np.zeros(len(dfp))
        plt.bar(xlabels, vals, bottom=bottoms, color=_cat_color(cat), edgecolor="white", linewidth=0.4, label=cat)
        bottoms += vals
    for i, v in enumerate(dfp["mean_b1"].values):
        plt.text(i, v+1.0, f"{v:.0f}", ha="center", va="bottom", fontsize=8.5, color="#333")
    plt.ylim(0, 110)
    plt.ylabel("Sub_B1 medio (0–100)", fontsize=10)
    plt.title("Series — Sub_B1 medio y composición de categorías", fontsize=11, color=EM_BLUE, pad=8)
    plt.xticks(rotation=0, fontsize=9)
    plt.grid(axis="y", alpha=0.25)
    plt.legend(ncol=min(6,len(cat_order)), fontsize=8, frameon=False, loc="upper center", bbox_to_anchor=(0.5, -0.12))
    plt.tight_layout(rect=[0, 0.04, 1, 1])
    p = os.path.join(out_dir, "resumen_series__SubB1medio_y_categorias.png")
    plt.savefig(p, bbox_inches="tight", dpi=150); plt.close()
    png_paths.append(p)

# 2.3 Series temporales por sitio (si hay ≥2 fechas)
series_candidates = subs.groupby("sitio")["fecha"].nunique()
multi_series_sites = sorted(series_candidates[series_candidates>1].index.tolist())
for sitio_i in multi_series_sites:
    s_sub = subs.loc[subs["sitio"]==sitio_i, ["fecha","Sub_B1"]].dropna().sort_values("fecha")
    s_ec  = datos.loc[datos["sitio"]==sitio_i, ["fecha","Ecoli_UFC100ml"]].copy()
    s_ec["Ecoli_UFC100ml"] = pd.to_numeric(s_ec["Ecoli_UFC100ml"], errors="coerce")
    s_ec = s_ec.dropna().sort_values("fecha")

    if not s_sub.empty:
        plt.figure(figsize=(7.0, 3.0))
        plt.plot(s_sub["fecha"], s_sub["Sub_B1"], marker="o")
        plt.title(f"Sub_B1 — {sitio_i}", fontsize=11, color=EM_BLUE)
        plt.xlabel("Fecha"); plt.ylabel("Sub_B1 (0–100)")
        plt.ylim(0,110); plt.grid(True, alpha=0.3)
        p = os.path.join(out_dir, f"serie_temporal__SubB1__{_slug(sitio_i)}.png")
        plt.savefig(p, bbox_inches="tight", dpi=150); plt.close()
        png_paths.append(p)

    if not s_ec.empty:
        plt.figure(figsize=(7.0, 3.0))
        plt.plot(s_ec["fecha"], s_ec["Ecoli_UFC100ml"], marker="o")
        plt.title(f"E. coli — {sitio_i}", fontsize=11, color=EM_BLUE)
        plt.xlabel("Fecha"); plt.ylabel("UFC/100 mL"); plt.yscale("log")
        plt.axhline(126, linestyle="--"); plt.axhline(410, linestyle="--")
        plt.grid(True, which="both", alpha=0.3)
        p = os.path.join(out_dir, f"serie_temporal__Ecoli__{_slug(sitio_i)}.png")
        plt.savefig(p, bbox_inches="tight", dpi=150); plt.close()
        png_paths.append(p)

_card(f"🖼️ <b>Gráficos</b> generados: {len(png_paths)} PNG(s).")

# ---------------------------------------------
# 3) Mapa HTML (última medición por sitio)
# ---------------------------------------------
map_path = None
try:
    import folium
    HAVE_FOL = True
except Exception:
    HAVE_FOL = False

def _latest_by_site(df, date_col="fecha"):
    return df.sort_values(["sitio", date_col]).groupby("sitio", as_index=False).tail(1)

def _build_map(_sitios, _latest):
    lat0 = _sitios["latitud"].astype(float).mean()
    lon0 = _sitios["longitud"].astype(float).mean()
    m = folium.Map(location=[lat0, lon0], zoom_start=11, control_scale=True)
    info = _latest[["sitio","fecha","Clasificacion_final","Sub_B1","Estado_B2"]]
    mdf = pd.merge(_sitios, info, on="sitio", how="left")
    for _, r in mdf.iterrows():
        try:
            lat, lon = float(r["latitud"]), float(r["longitud"])
        except Exception:
            continue
        col = _cat_color(r.get("Clasificacion_final","Indeterminada"))
        fecha_txt = pd.to_datetime(r.get("fecha")).date().isoformat() if pd.notna(r.get("fecha")) else "—"
        pop = (f"<b style='color:{EM_BLUE}'>{r.get('sitio','')}</b><br>"
               f"{(r.get('descripcion','') or '')}<br>"
               f"<b>Fecha:</b> {fecha_txt}<br>"
               f"<b>Sub_B1:</b> {r.get('Sub_B1',np.nan)}<br>"
               f"<b>Sanitario:</b> {r.get('Estado_B2','')}<br>"
               f"<b>Clasificación:</b> <span style='color:{col}; font-weight:700'>{r.get('Clasificacion_final','')}</span>")
        folium.CircleMarker(
            location=[lat, lon], radius=7,
            color=col, fill=True, fill_color=col, fill_opacity=0.9,
            popup=folium.Popup(pop, max_width=340)
        ).add_to(m)
    return m

if HAVE_FOL and isinstance(sitios, pd.DataFrame) and not sitios.empty and {"sitio","latitud","longitud"}.issubset(set(sitios.columns)):
    latest_small = _latest_by_site(clf)[["sitio","fecha","Clasificacion_final","Sub_B1","Estado_B2"]].copy()
    fmap_to_save = _build_map(sitios.copy(), latest_small)
    map_path = os.path.join(out_dir, "mapa_ecomonitor_wqi.html")
    fmap_to_save.save(map_path)
    _card(f"🗺️ <b>Mapa</b> exportado: <code>{map_path}</code>")
else:
    _card("🗺️ Mapa no exportado (faltan coordenadas en <code>df_sitios</code> o el paquete <code>folium</code>).")

# ---------------------------------------------
# 4) README.txt
# ---------------------------------------------
readme_path = os.path.join(out_dir, "README.txt")
ts = datetime.datetime.now().strftime("%Y-%m-%d %H:%M")
with open(readme_path, "w", encoding="utf-8") as f:
    f.write("EcoMonitor WQI — Paquete de resultados\n")
    f.write(f"Exportado: {ts}\n\n")
    f.write("Contenido:\n")
    f.write("- EcoMonitor_WQI_resultados.xlsx : tablas consolidadas\n")
    if png_paths:
        f.write("- PNGs: figuras por serie y series temporales\n")
    if map_path:
        f.write("- mapa_ecomonitor_wqi.html : mapa interactivo (última medición por sitio)\n")
    f.write("\nNotas:\n")
    f.write("- La clasificación final se basa en la matriz Bloque1×Bloque2 con veto sanitario.\n")
    f.write("- Umbrales recreativos definidos en 'umbrales' (Sección 3).\n")

_card(f"📝 <b>README</b> creado: <code>{readme_path}</code>")

# ---------------------------------------------
# 5) ZIP con todo + botones de descarga
# ---------------------------------------------
zip_path = os.path.join(out_dir, "EcoMonitor_WQI_paquete.zip")
with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as z:
    z.write(excel_path, arcname=os.path.basename(excel_path))
    for p in png_paths:
        if os.path.exists(p): z.write(p, arcname=os.path.basename(p))
    if map_path and os.path.exists(map_path):
        z.write(map_path, arcname=os.path.basename(map_path))
    z.write(readme_path, arcname=os.path.basename(readme_path))

_card(f"📦 <b>ZIP final listo</b>: <code>{zip_path}</code>")

# Botones (Colab friendly)
try:
    from google.colab import files
    _COLAB = True
except Exception:
    _COLAB = False

import ipywidgets as widgets
def _dl(path):
    try:
        files.download(path)
    except Exception:
        _card(f"Archivo disponible en <code>{path}</code>")

btn_zip = widgets.Button(description="📥 Descargar ZIP (todo)")
btn_zip.add_class("em-btn"); btn_zip.add_class("em-btn-green")
btn_zip.on_click(lambda _ : _dl(zip_path))

btn_xlsx = widgets.Button(description="📥 Descargar Excel consolidado")
btn_xlsx.add_class("em-btn"); btn_xlsx.add_class("em-btn-blue")
btn_xlsx.on_click(lambda _ : _dl(excel_path))

display(widgets.HBox([btn_zip, btn_xlsx]))

# Resumen compacto
display(HTML(
    f"<div style='margin:8px 0 0 0; color:#333;'>"
    f"<b>Resumen:</b> Excel=<{os.path.basename(excel_path)}>, PNGs={len(png_paths)}, "
    f"Mapa={'sí' if map_path else 'no'}, ZIP=<{os.path.basename(zip_path)}></div>"
))
```

```python
# @title
# =============================================
# Sección 7 — Informe PDF (Info del índice AL FINAL) + botón de descarga
# - Portada breve (estilo EcoMonitor)
# - Tabla de clasificación (todas las muestras) — sitio/fecha/serie primero (Sub_B1 redondeado)
# - Figuras clave (2): 1) Sub_B1 por sitio (última serie)  2) Series — Sub_B1 medio + composición
# - Nota sobre mapa (HTML)
# - ***Al final***: “Información del índice” (variables, bloque, sentido, unidad, objetivo, crítico, peso[2d], peso_norm[2d])
# =============================================
import os, sys, subprocess, io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as w
from IPython.display import display, HTML

# ---------- Estilo EcoMonitor ----------
EM_BLUE  = globals().get("EM_BLUE",  "#135566")
EM_GREEN = globals().get("EM_GREEN", "#7BC043")
EM_LIGHT = globals().get("EM_LIGHT", "#F2F7F4")
STYLE_BOX = globals().get(
    "STYLE_BOX",
    f"background:{EM_LIGHT}; border-left:4px solid {EM_GREEN}; padding:10px 12px; border-radius:10px;"
)

display(HTML(f"""
<style>
.em-btn.widget-button {{
  border-radius: 12px !important;
  font-weight: 700 !important;
  border: none !important;
  width: 300px !important; height: 40px !important;
  margin: 8px 0 4px 0 !important;
}}
.em-btn-blue.widget-button  {{ background: {EM_BLUE} !important; color:#fff !important; }}
.em-btn-green.widget-button {{ background: {EM_GREEN} !important; color:#fff !important; }}

/* Salida sin scroll interno */
.output_wrapper, .output, .output_area {{ height:auto !important; max-height:none !important; }}
.output_scroll {{ overflow-y: visible !important; height:auto !important; max-height:none !important; box-shadow:none !important; }}
</style>
"""))

def _card(html):
    display(HTML(f"<div style='{STYLE_BOX}; margin:12px 0 16px 0'>{html}</div>"))

# ---------- Chequeos mínimos ----------
missing = []
if 'df_clasificacion' not in globals(): missing.append("df_clasificacion (Sección 5)")
if 'subindices' not in globals():       missing.append("subindices (Sección 4)")
if 'df_datos' not in globals():         missing.append("df_datos (Sección 1)")
if 'ref_rec' not in globals() and 'config_b1b2' not in globals():
    missing.append("umbrales/pesos (Sección 3: ref_rec o config_b1b2)")
if missing:
    raise RuntimeError("Faltan insumos: " + ", ".join(missing))

# Bases
clf   = df_clasificacion.copy()
subs  = subindices.copy()
datos = df_datos.copy()
um    = (config_b1b2.copy() if 'config_b1b2' in globals() else ref_rec.copy())

for d in (clf, subs, datos):
    if 'fecha' in d.columns:
        d['fecha'] = pd.to_datetime(d['fecha'], errors='coerce')

# Unir 'serie' desde dataset si aún no está en clf/subs
merge_keys = [k for k in ['id','sitio','fecha'] if k in clf.columns and k in datos.columns]
if 'serie' not in clf.columns and merge_keys:
    clf = clf.merge(datos[merge_keys + ['serie']].drop_duplicates(), on=merge_keys, how='left')
if 'serie' not in subs.columns and merge_keys:
    subs = subs.merge(datos[merge_keys + ['serie']].drop_duplicates(), on=merge_keys, how='left')

# ---------- Helpers (norm/format) ----------
def _to_float(x):
    if pd.isna(x): return np.nan
    if isinstance(x, str):
        x = x.replace(",", ".").strip()
        if x == "": return np.nan
    return pd.to_numeric(x, errors="coerce")

def _norm_sentido(s):
    s = str(s or "").strip().lower()
    mp = {"≤":"<=","=<":"<=","≥":">=","=>":">=","range":"rango","between":"rango"}
    return mp.get(s, s)

def _infer_bloque(p):
    p = str(p).strip()
    if p.lower() == "ecoli_ufc100ml": return "B2"
    if p in ["As_ugL","Pb_ugL","Cd_ugL","Hg_ugL","F_mgL"]: return "B2"
    return "B1"

# ---------- Preparar tabla “Información del índice” (Sección 3) ----------
um = um.copy()
um.columns = [c.strip().lower() for c in um.columns]
for c in ["parametro","unidad","sentido","objetivo","objetivo_max","peso","bloque"]:
    if c not in um.columns: um[c] = np.nan
um["parametro"]    = um["parametro"].astype(str).str.strip()
um["unidad"]       = um["unidad"].astype(str).str.strip()
um["sentido"]      = um["sentido"].apply(_norm_sentido)
um["objetivo"]     = um["objetivo"].apply(_to_float)
um["objetivo_max"] = um["objetivo_max"].apply(_to_float)
um["peso"]         = um["peso"].apply(_to_float).fillna(0.0)
um["bloque"]       = um["bloque"].astype(str).str.upper().replace({"":np.nan})
um.loc[um["bloque"].isna(), "bloque"] = um.loc[um["bloque"].isna(), "parametro"].apply(_infer_bloque)

# Peso normalizado B1 (solo variables activas con peso>0)
m_b1 = (um["bloque"].str.upper()=="B1") & (um["sentido"]!="") & (um["peso"]>0)
sum_w = um.loc[m_b1, "peso"].sum()
um["peso_norm"] = np.where(m_b1 & (sum_w>0), um["peso"]/sum_w, np.nan)

vars_pdf = (um[["parametro","bloque","sentido","unidad","objetivo","objetivo_max","peso","peso_norm"]]
            .copy()
            .rename(columns={"parametro":"Variable","bloque":"Subíndice","sentido":"Sentido",
                             "unidad":"Unidad","objetivo":"Objetivo","objetivo_max":"Crítico",
                             "peso":"Peso","peso_norm":"Peso_norm"}))
vars_pdf["Subíndice"] = vars_pdf["Subíndice"].str.upper()

def _fmt_num(x):
    if pd.isna(x): return ""
    try:
        v = float(x)
        if abs(v)>=100: return f"{v:.0f}"
        if abs(v)>=10:  return f"{v:.1f}"
        if abs(v)>=1:   return f"{v:.2f}"
        return f"{v:.3f}"
    except:
        return str(x)

# pesos con 2 decimales
vars_pdf["Objetivo"]  = vars_pdf["Objetivo"].apply(_fmt_num)
vars_pdf["Crítico"]   = vars_pdf["Crítico"].apply(_fmt_num)
vars_pdf["Peso"]      = vars_pdf["Peso"].apply(lambda v: (f"{float(v):.2f}" if pd.notna(v) else ""))
vars_pdf["Peso_norm"] = vars_pdf["Peso_norm"].apply(lambda v: (f"{float(v):.2f}" if pd.notna(v) else "—"))

# Orden: B1 con peso>0, luego resto B1, luego B2
order_key = np.where(
    (vars_pdf["Subíndice"].eq("B1")) & (pd.to_numeric(vars_pdf["Peso"], errors="coerce").fillna(0)>0), 0,
    np.where(vars_pdf["Subíndice"].eq("B1"), 1, 2)
)
vars_pdf = vars_pdf.assign(_ok=order_key).sort_values(["_ok","Subíndice","Variable"]).drop(columns="_ok")

# ---------- Figuras clave (2) para el PDF (se guardan PNG pequeños) ----------
out_dir = "/content/EcoMonitor_WQI_exports"
os.makedirs(out_dir, exist_ok=True)

CAT_COLORS = globals().get("CAT_COLORS", {
    "Apta":"#2E7D32","Aceptable":"#7CB342","Condicional":"#F9A825",
    "Mala":"#F4511E","No apta":"#C62828","Indeterminada":"#9E9E9E",
})
def _cat_color(c): return CAT_COLORS.get(str(c), "#9E9E9E")

# 1) Sub_B1 por sitio — última serie (si no hay 'serie', usar YYYY-MM)
if "serie" not in subs.columns:
    subs["serie"] = subs["fecha"].dt.to_period("M").astype(str)
if "serie" not in clf.columns:
    clf["serie"] = clf["fecha"].dt.to_period("M").astype(str)

ser_order = subs.groupby("serie")["fecha"].max().sort_values().index.tolist()
ultima_serie = ser_order[-1] if ser_order else None

png1 = None
if ultima_serie is not None:
    tmp = subs.loc[subs["serie"]==ultima_serie, ["sitio","Sub_B1"]].copy()
    tmp["Sub_B1"] = pd.to_numeric(tmp["Sub_B1"], errors="coerce")
    tmp = tmp.dropna()
    if not tmp.empty:
        per_site = tmp.groupby("sitio", as_index=False)["Sub_B1"].mean().sort_values("Sub_B1", ascending=False)
        # >>> redondeo a enteros SOLO para la figura <<<
        per_site["Sub_B1"] = per_site["Sub_B1"].round(0)
        plt.figure(figsize=(6.6, 3.2))
        plt.bar(per_site["sitio"].astype(str), per_site["Sub_B1"].astype(float), color=EM_BLUE)
        plt.title(f"Sub_B1 por sitio — última serie: {ultima_serie}", fontsize=10, color=EM_BLUE, pad=6)
        plt.ylim(0, 110)
        plt.xticks(rotation=45, ha="right", fontsize=8.5)
        plt.grid(axis="y", alpha=0.3)
        plt.tight_layout()
        png1 = os.path.join(out_dir, "pdf_subb1_ultima_serie.png")
        plt.savefig(png1, dpi=150, bbox_inches="tight"); plt.close()

# 2) Series — Sub_B1 medio y composición
rows = []
cat_order = ["Apta","Aceptable","Condicional","Mala","No apta","Indeterminada"]
for s in ser_order:
    ssub = subs.loc[subs["serie"]==s, ["Sub_B1"]].copy()
    ssub["Sub_B1"] = pd.to_numeric(ssub["Sub_B1"], errors="coerce").dropna()
    if ssub.empty:
        continue
    mean_b1 = float(ssub["Sub_B1"].mean())
    cc = clf.loc[clf["serie"]==s, ["sitio","fecha","Clasificacion_final"]].dropna(subset=["sitio"])
    if not cc.empty:
        cc = cc.sort_values(["sitio","fecha"]).groupby("sitio", as_index=False).tail(1)
        total = cc["sitio"].nunique()
        counts = cc["Clasificacion_final"].value_counts()
        shares = {cat: (counts.get(cat, 0)/total if total>0 else 0.0) for cat in cat_order}
    else:
        shares = {cat: 0.0 for cat in cat_order}
    rows.append({"serie": s, "mean_b1": mean_b1, **{f"seg_{c}": shares[c]*mean_b1 for c in cat_order}})

png2 = None
if rows:
    dfp = pd.DataFrame(rows).set_index("serie")
    dfp = dfp.reindex([s for s in ser_order if s in dfp.index])
    plt.figure(figsize=(6.6, 3.2))
    bottoms = np.zeros(len(dfp))
    xlabels = dfp.index.astype(str).tolist()
    for cat in cat_order:
        col = f"seg_{cat}"
        vals = dfp[col].values if col in dfp.columns else np.zeros(len(dfp))
        plt.bar(xlabels, vals, bottom=bottoms, color=_cat_color(cat), edgecolor="white", linewidth=0.4, label=cat)
        bottoms += vals
    for i, val in enumerate(dfp["mean_b1"].values):
        plt.text(i, val + 1.0, f"{val:.0f}", ha="center", va="bottom", fontsize=8.5, color="#333")
    plt.ylim(0, 110)
    plt.ylabel("Sub_B1 medio (0–100)", fontsize=9)
    plt.title("Series — Sub_B1 medio y composición de categorías", fontsize=10, color=EM_BLUE, pad=6)
    plt.xticks(rotation=0, fontsize=8.5)
    plt.grid(axis="y", alpha=0.25)
    plt.legend(ncol=min(6, len(cat_order)), fontsize=7.5, frameon=False, loc="upper center", bbox_to_anchor=(0.5, -0.2))
    plt.tight_layout(rect=[0, 0.06, 1, 1])
    png2 = os.path.join(out_dir, "pdf_series_apiladas.png")
    plt.savefig(png2, dpi=150, bbox_inches="tight"); plt.close()

# ---------- ReportLab (PDF) ----------
try:
    from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, Image, PageBreak
    from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
    from reportlab.lib import colors
    from reportlab.lib.pagesizes import A4
    from reportlab.lib.utils import ImageReader
    from reportlab.pdfgen.canvas import Canvas
except Exception:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "reportlab"])
    from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, Image, PageBreak
    from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
    from reportlab.lib import colors
    from reportlab.lib.pagesizes import A4
    from reportlab.lib.utils import ImageReader
    from reportlab.pdfgen.canvas import Canvas

PAGE_W, PAGE_H = A4
LM, RM, TM, BM = 28, 28, 28, 28

# helpers de color
def _hex(hexstr): return colors.HexColor(hexstr)
def _tint(hexstr, w=0.90):
    """Mezcla color con blanco (w = proporción de blanco)."""
    c = colors.HexColor(hexstr)
    r = w*1.0 + (1.0-w)*c.red
    g = w*1.0 + (1.0-w)*c.green
    b = w*1.0 + (1.0-w)*c.blue
    return colors.Color(r,g,b)

styles = getSampleStyleSheet()
styleH  = ParagraphStyle("H1", parent=styles["Heading1"], textColor=_hex(EM_BLUE), spaceAfter=6, spaceBefore=0)
styleH2 = ParagraphStyle("H2", parent=styles["Heading2"], textColor=_hex(EM_BLUE), spaceAfter=6, spaceBefore=6)
styleN  = ParagraphStyle("N",  parent=styles["Normal"],  fontSize=9, leading=12)
styleCap= ParagraphStyle("Cap", parent=styles["Normal"], fontSize=8, textColor=colors.grey, leading=10)

logo_path = None
for p in ["/content/logo.png","/content/em_logo.png"]:
    if os.path.exists(p): logo_path = p; break

def _header_footer(canvas: Canvas, doc):
    canvas.saveState()
    w, h = A4
    # banda fina verde
    canvas.setFillColor(_hex(EM_GREEN))
    canvas.rect(doc.leftMargin, h-52, w-doc.leftMargin-doc.rightMargin, 2.2, stroke=0, fill=1)
    # línea azul
    canvas.setStrokeColor(_hex(EM_BLUE)); canvas.setLineWidth(0.6)
    canvas.line(doc.leftMargin, h-54, w-doc.rightMargin, h-54)
    # logo
    if logo_path and os.path.exists(logo_path):
        try: canvas.drawImage(logo_path, w-doc.rightMargin-72, h-50, width=64, height=20, preserveAspectRatio=True, mask='auto')
        except: pass
    # pie
    canvas.setFont("Helvetica", 8); canvas.setFillColor(colors.grey)
    canvas.drawRightString(w-doc.rightMargin, 20, f"Página {doc.page}")
    canvas.restoreState()

def _image_block(path, max_w):
    try:
        ir = ImageReader(path)
        iw, ih = ir.getSize()
        scale = min(1.0, max_w/iw)
        w = max_w; h = ih*scale
        return Image(path, width=w, height=h)
    except Exception:
        return None

# ---------- Construcción del PDF ----------
pdf_path = os.path.join(out_dir, "EcoMonitor_WQI_informe.pdf")

def build_pdf(_download=True):
    doc = SimpleDocTemplate(pdf_path, pagesize=A4, leftMargin=LM, rightMargin=RM, topMargin=TM, bottomMargin=BM)
    content_w = PAGE_W - LM - RM
    elements = []

    # Portada
    elements += [
        Paragraph("EcoMonitor WQI — Informe Técnico", styleH),
        Paragraph("Calidad de agua con fines recreativos", styleH2),
        Spacer(1, 8),
        Paragraph(
            "Marco metodológico: <b>Bloque 1</b> (físico–químico, 0–100 ponderado) + <b>Bloque 2</b> "
            "(sanitario por <i>E. coli</i> y tóxicos). Integración por matriz Bloque1×Bloque2 y "
            "veto por excedencias críticas (incluye STV de <i>E. coli</i>).", styleN),
        Spacer(1, 12),
    ]

    # Tabla de clasificación (sitio/fecha/serie primero)
    clf_pdf = clf.copy()
    if 'fecha' in clf_pdf.columns:
        clf_pdf['fecha'] = pd.to_datetime(clf_pdf['fecha'], errors='coerce').dt.date
    # >>> redondeo SOLO para visualización
    if 'Sub_B1' in clf_pdf.columns:
        clf_pdf["Sub_B1"] = pd.to_numeric(clf_pdf["Sub_B1"], errors="coerce").round(0).astype("Int64")

    lead = [c for c in ["sitio","fecha","serie"] if c in clf_pdf.columns]
    rest = [c for c in clf_pdf.columns if c not in lead]
    clf_pdf = clf_pdf[lead + rest]

    elements += [
        Paragraph("Tabla de clasificación (todas las muestras)", styleH2),
        Paragraph("Columnas: " + ", ".join(map(str, clf_pdf.columns.tolist())), styleCap),
        Spacer(1, 4)
    ]

    headers = clf_pdf.columns.tolist()
    data_rows = clf_pdf.values.tolist()

    # estilo de tabla con colores EcoMonitor
    header_bg = _hex(EM_LIGHT)
    grid_col  = _hex(EM_BLUE)
    zebra     = _tint(EM_LIGHT, 0.96)  # muy suave

    BLOCK = 36
    for s in range(0, len(data_rows), BLOCK):
        chunk = data_rows[s:s+BLOCK]
        tbl = Table([headers]+chunk, repeatRows=1)
        tbl.setStyle(TableStyle([
            ("BACKGROUND",(0,0),(-1,0), header_bg),
            ("TEXTCOLOR",(0,0),(-1,0), _hex(EM_BLUE)),
            ("GRID",(0,0),(-1,-1), 0.25, grid_col),
            ("FONTSIZE",(0,0),(-1,-1), 8),
            ("ALIGN",(0,0),(-1,0),"CENTER"),
            ("VALIGN",(0,0),(-1,-1),"MIDDLE"),
            ("ROWBACKGROUNDS",(0,1),(-1,-1), [zebra, colors.white]),
            ("TOPPADDING",(0,0),(-1,-1), 4),
            ("BOTTOMPADDING",(0,0),(-1,-1), 4),
        ]))
        elements += [tbl, Spacer(1, 8)]
        if s + BLOCK < len(data_rows):
            elements.append(PageBreak())

    # Figuras clave (solo 2)
    figs = []
    if png1 and os.path.exists(png1): figs.append(("Sub_B1 por sitio — última serie", png1))
    if png2 and os.path.exists(png2): figs.append(("Series — Sub_B1 medio y composición", png2))
    if figs:
        elements += [PageBreak(), Paragraph("Figuras clave", styleH2),
                     Paragraph("Selección acotada para foco en hallazgos principales.", styleCap),
                     Spacer(1, 6)]
        for title, path in figs:
            img = _image_block(path, content_w)
            if img:
                elements += [Paragraph(title, styleN), Spacer(1, 3), img, Spacer(1, 10)]

    # Nota sobre mapa
    elements.append(Paragraph(
        "Nota: el mapa interactivo se exporta como archivo HTML (<i>mapa_ecomonitor_wqi.html</i>) en la carpeta de exportaciones (Sección 6.6).",
        styleCap))
    elements.append(PageBreak())

    # *** AL FINAL *** — Información del índice
    elements += [
        Paragraph("Información del índice (umbrales y pesos)", styleH2),
        Paragraph("Variables, bloque de pertenencia, sentido, unidades, objetivos, críticos y pesos normalizados.", styleCap),
        Spacer(1, 6),
    ]
    headers_vars = ["Variable","Subíndice","Sentido","Unidad","Objetivo","Crítico","Peso","Peso_norm"]
    rows_vars = vars_pdf[headers_vars].fillna("").values.tolist()

    STEP = 30
    for i in range(0, len(rows_vars), STEP):
        chunk = rows_vars[i:i+STEP]
        t = Table([headers_vars] + chunk, repeatRows=1)
        t.setStyle(TableStyle([
            ("BACKGROUND",(0,0),(-1,0), header_bg),
            ("TEXTCOLOR",(0,0),(-1,0), _hex(EM_BLUE)),
            ("GRID",(0,0),(-1,-1), 0.25, grid_col),
            ("FONTSIZE",(0,0),(-1,-1), 8),
            ("ALIGN",(0,0),(-1,0),"CENTER"),
            ("VALIGN",(0,0),(-1,-1),"MIDDLE"),
            ("ROWBACKGROUNDS",(0,1),(-1,-1), [zebra, colors.white]),
            ("TOPPADDING",(0,0),(-1,-1), 4),
            ("BOTTOMPADDING",(0,0),(-1,-1), 4),
        ]))
        elements += [t, Spacer(1, 8)]

    # Compilar
    doc.build(elements, onFirstPage=_header_footer, onLaterPages=_header_footer)

    # Descargar/avisar
    try:
        from google.colab import files
        if _download: files.download(pdf_path)
    except Exception:
        pass
    _card(f"📑 Informe PDF generado en:<br><code>{pdf_path}</code>")

# ---------- Botón (descargar informe) ----------
btn = w.Button(description="📄 Generar y descargar informe PDF")
btn.add_class("em-btn"); btn.add_class("em-btn-blue")
btn.on_click(lambda _ : build_pdf(_download=True))
display(btn)

# También podés construir sin descarga automática:
# build_pdf(_download=False)
```
